# 1️⃣ Introduction

A **Series** is a one-dimensional labeled array provided by the Pandas library.

It can store:

- Integers
- Floating-point numbers
- Strings
- Boolean values
- Dates
- Python objects

Unlike a normal Python list, every value in a Series has an associated **index**, making data retrieval faster and more organized.

You can think of a Series as a single column in an Excel sheet or a single column of a database table.

# 3️⃣ What is a Pandas Series?

A **Series** is a one-dimensional data structure that stores:

- Values
- Corresponding Index

Every element in a Series has:

1. A value
2. An index

Example:

| Index | Value |
|------|------|
|0|92|
|1|87|
|2|78|
|3|95|
|4|84|

The index acts like the address of each value, allowing quick and efficient access.

# 7️⃣ Creating a Series

Pandas allows us to create a Series from various Python data structures.

Common sources include:

- List
- Tuple
- Dictionary
- NumPy Array
- Scalar Value

Let's explore each one with examples.

## 7.1 Creating a Series from a Python List

A Python list is the most common way to create a Series.

By default, Pandas assigns integer indexes starting from 0.

````python
import pandas as pd

marks = [92, 87, 78, 95, 84]

series_marks = pd.Series(marks)

series_marks
````

### Explanation

- Left side → Index
- Right side → Values
- `dtype` → Data type of stored values

## Why use a Series instead of a List?

Python List

```python
marks = [92,87,78,95]
```

Pandas Series

```python
marks = pd.Series([92,87,78,95])
```

Series provides many built-in methods such as:

- mean()
- max()
- min()
- sum()
- count()
- sort_values()

These operations would require additional loops or functions when using a plain Python list.

## 7.2 Creating a Series from a Tuple

A tuple is immutable, but Pandas can still convert it into a Series.

````python
numbers = (10, 20, 30, 40)

pd.Series(numbers)
````

### Explanation

Although tuples cannot be modified, the created Series behaves like a normal Pandas Series.

The original tuple remains unchanged.

## 7.3 Creating a Series from a Dictionary

When creating a Series from a dictionary:

- Keys become the index.
- Values become the data.

````python
students = {
    "Rahul":92,
    "Priya":87,
    "Amit":78,
    "Sneha":95
}

student_marks = pd.Series(students)

student_marks
````

## Industry Usage

This method is useful when data already exists in key-value format.

Examples:

- Employee ID → Salary
- Product ID → Price
- Student ID → Marks

## 7.4 Creating a Series from a NumPy Array

Since Pandas is built on top of NumPy, NumPy arrays can easily be converted into a Series.

````python
import numpy as np

array = np.array([100,200,300,400])

pd.Series(array)
````

### Why is this useful?

Many Machine Learning libraries produce NumPy arrays.

Pandas allows us to convert them into Series for easier analysis.

## 7.5 Creating a Series from a Scalar Value

A scalar value is a single value.

When a scalar is used, the index must be specified.

````python
pd.Series(100, index=[0,1,2,3,4])
````

# 8️⃣ Custom Index

By default, Pandas assigns indexes automatically.

We can also specify our own indexes.

````python
marks = [92,87,78,95]

students = ["Rahul","Priya","Amit","Sneha"]

series = pd.Series(marks,index=students)

series
````

## Why are Custom Indexes Useful?

Examples:

Customer ID

Employee ID

Product ID

Roll Number

Country Name

These are much more meaningful than numeric indexes.

# 9️⃣ Understanding Series Attributes

Once a Series is created, Pandas provides several built-in attributes that help us understand its structure.

These attributes allow us to answer questions such as:

- What type of data is stored?
- How many elements are present?
- What are the indexes?
- What is the shape of the Series?
- Is it one-dimensional?

Understanding these attributes is an essential first step before performing any data analysis.

````python
import pandas as pd

students = pd.Series(
    [92, 87, 78, 95, 84],
    index=["Rahul", "Priya", "Amit", "Sneha", "Kiran"],
    name="Marks"
)

students
````

# 9.1 index

The **index** attribute returns the labels associated with each value in the Series.

Think of the index as the "address" of every value.

Unlike Python lists, the indexes can be customized to meaningful labels such as employee IDs, product IDs, or student names.

````python
students.index
````

### Expected Output

```
Index(['Rahul', 'Priya', 'Amit', 'Sneha', 'Kiran'], dtype='object')
```

### Explanation

The output shows the labels (indexes) assigned to each value.

These indexes help retrieve values quickly without relying only on numeric positions.

### Real Industry Usage

Instead of numeric indexes, industries often use:

- Employee IDs
- Customer IDs
- Product IDs
- Roll Numbers
- Transaction IDs

Meaningful indexes improve readability and make data analysis easier.

# 9.2 values

The **values** attribute returns the actual data stored in the Series as a NumPy array.

This is useful when you need to perform NumPy operations or interact with libraries that expect arrays.

````python
students.values
````

# 9.3 dtype

The **dtype** attribute tells us the data type of the values stored inside the Series.

Knowing the data type is important because many operations depend on it.

````python
students.dtype
````

### Expected Output

```
dtype('int64')
```

### Explanation

All values are integers.

Therefore, Pandas stores them as **int64**.

Other possible data types include:

- float64
- object
- bool
- datetime64

# 9.4 shape

The **shape** attribute tells us the dimensions of the Series.

Since a Series is one-dimensional, the shape contains only one value.

````python
students.shape
````

### Expected Output

```
(5,)
```

### Explanation

The Series contains **5 elements**.

The comma indicates that this is a one-dimensional structure.

# 9.5 size

The **size** attribute returns the total number of elements present in the Series.

````python
students.size
````

### Expected Output

```
5
```

Unlike `shape`, `size` returns only the number of elements.

# 9.6 ndim

The **ndim** attribute returns the number of dimensions of the Series.

````python
students.ndim
````

### Expected Output

```
1
```

### Explanation

A Series always has **one dimension**.

A DataFrame, on the other hand, has **two dimensions**.

# 9.7 name

The **name** attribute returns the name assigned to the Series.

Giving meaningful names improves readability, especially when multiple Series are combined into a DataFrame.

````python
students.name
````

### Expected Output

```
'Marks'
```

The name of this Series is **Marks**, which we assigned while creating it.

## Difference Between `shape` and `ndim`

### `ndim`

**Question:**

> How many dimensions does this object have?

Example:

```python
students.ndim
```

Output

```python
1
```

Meaning:

The Series has only **one dimension**.

---

### `shape`

**Question:**

> What is the size of each dimension?

Example:

```python
students.shape
```

Output

```python

```

Meaning:

---

### Comparison

| Attribute | Question it Answers | Example |
|------------|--------------------|----------|

| `shape` | What is the size of each dimension? | `(5,)` |
| `size` | Total number of elements? | `5` |

---

### 🧠 Memory Trick

Remember:

- **ndim → Number of Dimensions**
- **shape → Size of each Dimension**
- **size → Total number of values**

# 🔟 Accessing Data in a Series

Creating a Series is only the first step.

To analyze data effectively, we need to retrieve specific values from the Series.

Pandas provides multiple ways to access data:

- Square Brackets `[]`
- Label-based Indexing `.loc[]`
- Position-based Indexing `.iloc[]`

Choosing the correct method improves code readability and helps avoid common indexing mistakes.

````python
import pandas as pd

students = pd.Series(
    [92,87,78,95,84],
    index=["Rahul","Priya","Amit","Sneha","Kiran"],
    name="Marks"
)

students
````

# 10.1 Accessing Data Using Square Brackets `[]`

Square brackets are the simplest way to retrieve data from a Series.

Depending on the type of index used, they can retrieve values by:

- Label
- Position (when default integer index is present)

However, for clarity and consistency, `.loc[]` and `.iloc[]` are generally preferred.

````python
students["Rahul"]
````

Index

Rahul

↓

Value

92

````python
students["Sneha"]
````

## Why use labels instead of positions?

Imagine your Series contains employee salaries.

Accessing:

```python
salary["EMP1023"]
```

is much more meaningful than:

```python
salary[57]
```

Meaningful labels improve code readability.

# 10.2 Accessing Multiple Values

````python
students[["Rahul","Kiran","Priya"]]
````

Explanation

Passing a list of labels returns another Series containing only those values.

# 10.3 Accessing Using Position (iloc)

`.iloc[]` stands for **Integer Location**.

It accesses values based on their **position**, starting from **0**.

Use `.iloc[]` when you know the position of the data instead of its label.

Position

0 → Rahul

1 → Priya

2 → Amit

3 → Sneha

4 → Kiran

````python
students.iloc[0]
````

````python
students.iloc[3]
````

# 10.4 Accessing Using Labels (loc)

`.loc[]` stands for **Location**.

It retrieves data using the **index label**, not the position.

Whenever your Series has meaningful indexes such as names or IDs, `.loc[]` is the recommended approach.

````python
students.loc["Amit"]
````

````python
students.loc["Priya"]
````

| `.loc[]`      | `.iloc[]`         |
| ------------- | ----------------- |
| Uses Labels   | Uses Positions    |
| Employee ID   | Employee Number   |
| Student Name  | Row Number        |
| String Labels | Integer Positions |


# 10.5 Negative Indexing

Just like Python lists, .iloc[] supports negative indexing.

````python
students.iloc[-1]
````

# 1️⃣1️⃣ Updating Values in a Series

A Series is mutable, which means its values can be modified after creation.

Pandas allows updating values using:

- Label (`.loc[]`)
- Position (`.iloc[]`)
- Square Brackets (`[]`)

Updating values is commonly required while cleaning datasets, correcting mistakes, or replacing outdated information.

````python
import pandas as pd

students = pd.Series(
    [92,87,78,95,84],
    index=["Rahul","Priya","Amit","Sneha","Kiran"],
    name="Marks"
)

students
````

## 11.1 Updating Using Label

When we know the label (index), we can update the value directly.

This is the safest and most readable approach.

````python
students["Rahul"] = 98

students
````

# Industry Usage 

Suppose a student's marks were entered incorrectly.

Instead of recreating the Series, we simply update the value.

Similarly,

- Updating employee salaries
- Correcting product prices
- Fixing sensor readings

## 11.2 Updating Using `.loc[]`

`.loc[]` updates values using index labels.

This is the recommended approach when working with meaningful labels.

````python
students.loc["Priya"] = 91

students
````

## 11.3 Updating Using `.iloc[]`

`.iloc[]` updates values using integer positions.

Use this when the row position is known.

````python
students.iloc[2] = 80

students
````

## 11.4 Updating Multiple Values

````python
students[["Rahul","Sneha"]] = [100,99]

students
````

### How does this work?

The first value (100) is assigned to Rahul.

The second value (99) is assigned to Sneha.

The number of new values must match the number of selected indexes.

## ⚠️ Common Beginner Mistake

````python
# students[["Rahul","Sneha"]] = [100]

# ValueError:
# cannot set using a list-like indexer with a different length than the value
````

Selected

↓

2 Students

Given

↓

1 Value

Pandas doesn't know what value to assign to the second student.

````python
# Correct Solution

students[["Rahul","Sneha"]] = [100,99]
````

## Pros

✔ Easy to modify data

✔ Cleaner than loops

✔ Supports label and position based updates

✔ Useful during data cleaning

## Cons

❌ Accidentally overwrites existing data

❌ Wrong labels create new rows in some situations

❌ Position changes may affect updates

## Edge Cases

````python
students["Arjun"] = 88

students
````

# Explanation

Pandas automatically creates a new index.

Many beginners expect an error.

Instead,

Pandas adds a new entry.
