# 1️⃣ Introduction

A **Series** is a one-dimensional labeled array provided by the Pandas library.

It can store:

- Integers
- Floating-point numbers
- Strings
- Boolean values
- Dates
- Python objects

Unlike a normal Python list, every value in a Series has an associated **index**, making data retrieval faster and more organized.

You can think of a Series as a single column in an Excel sheet or a single column of a database table.

# 7️⃣ Creating a Series

## 7.1 Creating a Series from a Python List

A Python list is the most common way to create a Series.

By default, Pandas assigns integer indexes starting from 0.

````python
import pandas as pd

marks = [92, 87, 78, 95, 84]

series_marks = pd.Series(marks)

series_marks
````

### Explanation

- Left side → Index
- Right side → Values
- `dtype` → Data type of stored values

## Why use a Series instead of a List?

Python List

```python
marks = [92,87,78,95]
```

Pandas Series

```python
marks = pd.Series([92,87,78,95])
```

Series provides many built-in methods such as:

- mean()
- max()
- min()
- sum()
- count()
- sort_values()

These operations would require additional loops or functions when using a plain Python list.

## 7.2 Creating a Series from a Tuple

A tuple is immutable, but Pandas can still convert it into a Series.

````python
numbers = (10, 20, 30, 40)

pd.Series(numbers)
````

### Explanation

Although tuples cannot be modified, the created Series behaves like a normal Pandas Series.

The original tuple remains unchanged.

## 7.3 Creating a Series from a Dictionary

When creating a Series from a dictionary:

- Keys become the index.
- Values become the data.

````python
students = {
    "Rahul":92,
    "Priya":87,
    "Amit":78,
    "Sneha":95
}

student_marks = pd.Series(students)

student_marks
````

## Industry Usage

This method is useful when data already exists in key-value format.

Examples:

- Employee ID → Salary
- Product ID → Price
- Student ID → Marks

## 7.4 Creating a Series from a NumPy Array

Since Pandas is built on top of NumPy, NumPy arrays can easily be converted into a Series.

````python
import numpy as np

array = np.array([100,200,300,400])

pd.Series(array)
````

### Why is this useful?

Many Machine Learning libraries produce NumPy arrays.

Pandas allows us to convert them into Series for easier analysis.

## 7.5 Creating a Series from a Scalar Value

A scalar value is a single value.

When a scalar is used, the index must be specified.

````python
pd.Series(100, index=[0,1,2,3,4])
````

# 🔟 Accessing Data in a Series

Creating a Series is only the first step.

To analyze data effectively, we need to retrieve specific values from the Series.

Pandas provides multiple ways to access data:

- Square Brackets `[]`
- Label-based Indexing `.loc[]`
- Position-based Indexing `.iloc[]`

Choosing the correct method improves code readability and helps avoid common indexing mistakes.

````python
import pandas as pd

students = pd.Series(
    [92,87,78,95,84],
    index=["Rahul","Priya","Amit","Sneha","Kiran"],
    name="Marks"
)

students
````

# 10.1 Accessing Data Using Square Brackets `[]`

Square brackets are the simplest way to retrieve data from a Series.

Depending on the type of index used, they can retrieve values by:

- Label
- Position (when default integer index is present)

However, for clarity and consistency, `.loc[]` and `.iloc[]` are generally preferred.

````python
students["Rahul"]
````

Index

Rahul

↓

Value

92

````python
students["Sneha"]
````

## Why use labels instead of positions?

Imagine your Series contains employee salaries.

Accessing:

```python
salary["EMP1023"]
```

is much more meaningful than:

```python
salary[57]
```

Meaningful labels improve code readability.

# 10.2 Accessing Multiple Values

````python
students[["Rahul","Kiran","Priya"]]
````

Explanation

Passing a list of labels returns another Series containing only those values.

# 10.3 Accessing Using Position (iloc)

`.iloc[]` stands for **Integer Location**.

It accesses values based on their **position**, starting from **0**.

Use `.iloc[]` when you know the position of the data instead of its label.

Position

0 → Rahul

1 → Priya

2 → Amit

3 → Sneha

4 → Kiran

````python
students.iloc[0]
````

````python
students.iloc[3]
````

# 10.4 Accessing Using Labels (loc)

`.loc[]` stands for **Location**.

It retrieves data using the **index label**, not the position.

Whenever your Series has meaningful indexes such as names or IDs, `.loc[]` is the recommended approach.

````python
students.loc["Amit"]
````

````python
students.loc["Priya"]
````

| `.loc[]`      | `.iloc[]`         |
| ------------- | ----------------- |
| Uses Labels   | Uses Positions    |
| Employee ID   | Employee Number   |
| Student Name  | Row Number        |
| String Labels | Integer Positions |

# 1️⃣1️⃣ Updating Values in a Series

A Series is mutable, which means its values can be modified after creation.

Pandas allows updating values using:

- Label (`.loc[]`)
- Position (`.iloc[]`)
- Square Brackets (`[]`)

Updating values is commonly required while cleaning datasets, correcting mistakes, or replacing outdated information.

````python
import pandas as pd

students = pd.Series(
    [92,87,78,95,84],
    index=["Rahul","Priya","Amit","Sneha","Kiran"],
    name="Marks"
)

students
````

## 11.1 Updating Using Label

When we know the label (index), we can update the value directly.

This is the safest and most readable approach.

````python
students["Rahul"] = 98

students
````

# Industry Usage 

Suppose a student's marks were entered incorrectly.

Instead of recreating the Series, we simply update the value.

Similarly,

- Updating employee salaries
- Correcting product prices
- Fixing sensor readings

## 11.2 Updating Using `.loc[]`

`.loc[]` updates values using index labels.

This is the recommended approach when working with meaningful labels.

````python
students.loc["Priya"] = 91

students
````

## 11.3 Updating Using `.iloc[]`

`.iloc[]` updates values using integer positions.

Use this when the row position is known.

````python
students.iloc[2] = 80

students
````

## 11.4 Updating Multiple Values

````python
students[["Rahul","Sneha"]] = [100,99]

students
````

### How does this work?

The first value (100) is assigned to Rahul.

The second value (99) is assigned to Sneha.

The number of new values must match the number of selected indexes.

## ⚠️ Common Beginner Mistake

````python
# students[["Rahul","Sneha"]] = [100]

# ValueError:
# cannot set using a list-like indexer with a different length than the value
````

Selected

↓

2 Students

Given

↓

1 Value

Pandas doesn't know what value to assign to the second student.

````python
# Correct Solution

students[["Rahul","Sneha"]] = [100,99]
````

## Pros

✔ Easy to modify data

✔ Cleaner than loops

✔ Supports label and position based updates

✔ Useful during data cleaning

## Cons

❌ Accidentally overwrites existing data

❌ Wrong labels create new rows in some situations

❌ Position changes may affect updates

## Edge Cases

````python
students["Arjun"] = 88

students
````

# Explanation

Pandas automatically creates a new index.

Many beginners expect an error.

Instead,

Pandas adds a new entry.

# 12.4 count()

## 📖 Definition

The `count()` method returns the total number of **non-missing (non-null)** values present in a Series.

Unlike Python's `len()`, `count()` ignores missing values (`NaN`).

---

## 🤔 Why do we need it?

Before performing any analysis, we should know how many valid observations are available.

This helps us:

- Check data completeness
- Detect missing values
- Verify the dataset before applying Machine Learning algorithms

---

## 🧠 Think Like a Data Analyst

Imagine you collected marks of 100 students.

Some students were absent, so their marks are missing (`NaN`).

If you calculate the average, you should only consider students who actually have marks.

`count()` tells you exactly how many valid records are available.

````python
import pandas as pd
import numpy as np

students = pd.Series(
    [92, 87, 78, 95, np.nan],
    index=["Rahul", "Priya", "Amit", "Sneha", "Kiran"],
    name="Marks"
)

students
````

````python
students.count()
````

````python
len(students)
````

### 📤 Expected Output

```
4
```

---

### 🔍 Explanation

Our Series contains **5 students**.

However,

Kiran's marks are missing (`NaN`).

`count()` ignores missing values.

Therefore,

Valid Marks = **4**


### ⚡ Difference Between `count()` and `len()`

| `count()` | `len()` |
|------------|----------|
| Counts only non-null values | Counts all values |
| Ignores NaN | Includes NaN |
| Used in Data Analysis | Python built-in function |


## ⚠️ Common Beginner Mistake

Many beginners think:

```python
students.count()
```

and

```python
len(students)
```

always return the same answer.

This is true **only if there are no missing values**.

````python
students = pd.Series([10,20,np.nan,40])

print("count():", students.count())
print("len():", len(students))
````

## 🏭 Real Industry Usage

Suppose a hospital dataset contains patient blood pressure readings.

Some patients missed their check-up.

Before calculating statistics, analysts first check:

```python
patients["BP"].count()
```

to know how many valid readings are available.

---

## 👍 Pros

- Ignores missing values automatically
- Fast
- Very useful in EDA

---

## 👎 Cons

- Doesn't tell where missing values are
- Doesn't count NaN values

---

## ⚠️ Edge Case

```python

```

```

```

# 12.6 mean()

## 📖 Definition

The `mean()` method returns the arithmetic average of all numeric values in the Series.

By default, it ignores missing values (`NaN`).

---

## 🤔 Why do we need it?

The average helps us understand the overall trend of the data.

It is one of the most commonly used statistical measures in:

- Data Analysis
- Machine Learning
- Business Intelligence
- Finance
- Healthcare

---

## 🧠 Think Like a Data Scientist

Suppose a teacher wants to know whether an exam was difficult.

Instead of checking every student's marks, the teacher calculates the class average.

That's exactly what `mean()` does.

````python
students.mean()
````

### 📤 Expected Output

```
88.0
```

---

### 🔍 Explanation

```
(92 + 87 + 78 + 95)

/

4

=

88.0
```

The missing value (`NaN`) is ignored.

## ⚠️ Common Beginner Mistake

Many beginners think the average is calculated using all rows.

Actually,

Pandas divides only by the number of **valid (non-null)** values.

````python
print("Sum:", students.sum())
print("Count:", students.count())
print("Mean:", students.mean())
````

### Output

```
Sum: 352

Count: 4

Mean: 88.0
```

Notice the relationship:

```
Mean

=

Sum

÷

Count
```

This is a great way to verify whether your calculations are correct.

## 🏭 Real Industry Usage

The `mean()` method is used to calculate:

- Average student marks
- Average customer spending
- Average employee salary
- Average monthly sales
- Average temperature
- Average patient heart rate

## 👍 Pros

- Easy to interpret
- Widely used in statistics and ML
- Ignores missing values automatically

---

## 👎 Cons

- Sensitive to outliers
- Not suitable for highly skewed data

---

## ⚠️ Edge Case

If every value is `NaN`

```python
pd.Series([np.nan,np.nan]).mean()
```

Output

```
NaN
```

## 🧠 Interview Question

**Q:** How is `mean()` calculated in Pandas?

**Answer:**

```
Mean = Sum of non-null values ÷ Count of non-null values
```
