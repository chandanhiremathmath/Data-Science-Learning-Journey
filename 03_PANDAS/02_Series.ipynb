# 1️⃣ Introduction

A **Series** is a one-dimensional labeled array provided by the Pandas library.

It can store:

- Integers
- Floating-point numbers
- Strings
- Boolean values
- Dates
- Python objects

Unlike a normal Python list, every value in a Series has an associated **index**, making data retrieval faster and more organized.

You can think of a Series as a single column in an Excel sheet or a single column of a database table.

# 3️⃣ What is a Pandas Series?

A **Series** is a one-dimensional data structure that stores:

- Values
- Corresponding Index

Every element in a Series has:

1. A value
2. An index

Example:

| Index | Value |
|------|------|
|0|92|
|1|87|
|2|78|
|3|95|
|4|84|

The index acts like the address of each value, allowing quick and efficient access.

# 💡 Real-World Analogy

Imagine your college attendance register.

| Roll No | Student |
|----------|----------|
|101|Rahul|
|102|Priya|
|103|Amit|
|104|Sneha|

Here,

- Roll Number acts as the **Index**
- Student Name acts as the **Value**

A Pandas Series works in exactly the same way.

# 7️⃣ Creating a Series

Pandas allows us to create a Series from various Python data structures.

Common sources include:

- List
- Tuple
- Dictionary
- NumPy Array
- Scalar Value

Let's explore each one with examples.

## 7.1 Creating a Series from a Python List

A Python list is the most common way to create a Series.

By default, Pandas assigns integer indexes starting from 0.

````python
import pandas as pd

marks = [92, 87, 78, 95, 84]

series_marks = pd.Series(marks)

series_marks
````

### Explanation

- Left side → Index
- Right side → Values
- `dtype` → Data type of stored values

## Why use a Series instead of a List?

Python List

```python
marks = [92,87,78,95]
```

Pandas Series

```python
marks = pd.Series([92,87,78,95])
```

Series provides many built-in methods such as:

- mean()
- max()
- min()
- sum()
- count()
- sort_values()

These operations would require additional loops or functions when using a plain Python list.

## 7.2 Creating a Series from a Tuple

A tuple is immutable, but Pandas can still convert it into a Series.

````python
numbers = (10, 20, 30, 40)

pd.Series(numbers)
````

### Explanation

Although tuples cannot be modified, the created Series behaves like a normal Pandas Series.

The original tuple remains unchanged.

## 7.3 Creating a Series from a Dictionary

When creating a Series from a dictionary:

- Keys become the index.
- Values become the data.

````python
students = {
    "Rahul":92,
    "Priya":87,
    "Amit":78,
    "Sneha":95
}

student_marks = pd.Series(students)

student_marks
````

## Industry Usage

This method is useful when data already exists in key-value format.

Examples:

- Employee ID → Salary
- Product ID → Price
- Student ID → Marks

## 7.4 Creating a Series from a NumPy Array

Since Pandas is built on top of NumPy, NumPy arrays can easily be converted into a Series.

````python
import numpy as np

array = np.array([100,200,300,400])

pd.Series(array)
````

### Why is this useful?

Many Machine Learning libraries produce NumPy arrays.

Pandas allows us to convert them into Series for easier analysis.

## 7.5 Creating a Series from a Scalar Value

A scalar value is a single value.

When a scalar is used, the index must be specified.

````python
pd.Series(100, index=[0,1,2,3,4])
````

# 8️⃣ Custom Index

By default, Pandas assigns indexes automatically.

We can also specify our own indexes.

````python
marks = [92,87,78,95]

students = ["Rahul","Priya","Amit","Sneha"]

series = pd.Series(marks,index=students)

series
````

## Why are Custom Indexes Useful?

Examples:

Customer ID

Employee ID

Product ID

Roll Number

Country Name

These are much more meaningful than numeric indexes.

# 9️⃣ Understanding Series Attributes

Once a Series is created, Pandas provides several built-in attributes that help us understand its structure.

These attributes allow us to answer questions such as:

- What type of data is stored?
- How many elements are present?
- What are the indexes?
- What is the shape of the Series?
- Is it one-dimensional?

Understanding these attributes is an essential first step before performing any data analysis.

````python
import pandas as pd

students = pd.Series(
    [92, 87, 78, 95, 84],
    index=["Rahul", "Priya", "Amit", "Sneha", "Kiran"],
    name="Marks"
)

students
````

# 9.1 index

The **index** attribute returns the labels associated with each value in the Series.

Think of the index as the "address" of every value.

Unlike Python lists, the indexes can be customized to meaningful labels such as employee IDs, product IDs, or student names.

````python
students.index
````

### Expected Output

```
Index(['Rahul', 'Priya', 'Amit', 'Sneha', 'Kiran'], dtype='object')
```

### Explanation

The output shows the labels (indexes) assigned to each value.

These indexes help retrieve values quickly without relying only on numeric positions.

### Real Industry Usage

Instead of numeric indexes, industries often use:

- Employee IDs
- Customer IDs
- Product IDs
- Roll Numbers
- Transaction IDs

Meaningful indexes improve readability and make data analysis easier.

# 9.2 values

The **values** attribute returns the actual data stored in the Series as a NumPy array.

This is useful when you need to perform NumPy operations or interact with libraries that expect arrays.

````python
students.values
````

# 9.3 dtype

The **dtype** attribute tells us the data type of the values stored inside the Series.

Knowing the data type is important because many operations depend on it.

````python
students.dtype
````

### Expected Output

```
dtype('int64')
```

### Explanation

All values are integers.

Therefore, Pandas stores them as **int64**.

Other possible data types include:

- float64
- object
- bool
- datetime64

# 9.4 shape

The **shape** attribute tells us the dimensions of the Series.

Since a Series is one-dimensional, the shape contains only one value.

````python
students.shape
````

### Expected Output

```
(5,)
```

### Explanation

The Series contains **5 elements**.

The comma indicates that this is a one-dimensional structure.

# 9.5 size

The **size** attribute returns the total number of elements present in the Series.

````python
students.size
````

### Expected Output

```
5
```

Unlike `shape`, `size` returns only the number of elements.

# 9.6 ndim

The **ndim** attribute returns the number of dimensions of the Series.

````python
students.ndim
````

### Expected Output

```
1
```

### Explanation

A Series always has **one dimension**.

A DataFrame, on the other hand, has **two dimensions**.

# 9.7 name

The **name** attribute returns the name assigned to the Series.

Giving meaningful names improves readability, especially when multiple Series are combined into a DataFrame.

````python
students.name
````

### Expected Output

```
'Marks'
```

The name of this Series is **Marks**, which we assigned while creating it.

# 📊 Summary of Series Attributes

| Attribute | Description | Returns |
|------------|-------------|----------|
| index | Labels of the Series | Index Object |
| values | Data stored in the Series | NumPy Array |
| dtype | Data type of values | Data Type |
| shape | Dimensions | Tuple |
| size | Number of elements | Integer |
| ndim | Number of dimensions | Integer |
| name | Name of the Series | String |

## Difference Between `shape` and `ndim`

### `ndim`

**Question:**

> How many dimensions does this object have?

Example:

```python
students.ndim
```

Output

```python
1
```

Meaning:

The Series has only **one dimension**.

---

### `shape`

**Question:**

> What is the size of each dimension?

Example:

```python
students.shape
```

Output

```python
(5,)
```

Meaning:

There is **1 dimension**, and it contains **5 elements**.

---

### Comparison

| Attribute | Question it Answers | Example |
|------------|--------------------|----------|
| `ndim` | How many dimensions? | `1` |
| `shape` | What is the size of each dimension? | `(5,)` |
| `size` | Total number of elements? | `5` |

---

### 🧠 Memory Trick

Remember:

- **ndim → Number of Dimensions**
- **shape → Size of each Dimension**
- **size → Total number of values**
