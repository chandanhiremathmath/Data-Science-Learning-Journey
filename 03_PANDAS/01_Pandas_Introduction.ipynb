- Read datasets
- Clean datasets
- Transform data
- Analyze data

# 3️⃣ Why was Pandas Created?

Before Pandas, Python programmers mainly used:

- Lists
- Tuples
- Dictionaries

Although these data structures are powerful, they become difficult to manage when working with large datasets.

For example:

Imagine handling a dataset with 1 million rows using Python lists.

Tasks such as:

- Filtering rows
- Sorting values
- Calculating averages
- Removing duplicates
- Finding missing values

would require writing many loops and custom logic.

Pandas was developed to solve these problems efficiently.

````python
# Importing Pandas

import pandas as pd
````

## Why do we write "as pd"?

"Pandas" is a long library name.

Instead of writing:

pandas.read_csv()

the Data Science community follows a standard convention:

pd.read_csv()

This improves readability and keeps the code concise.

Although you can choose any alias, "pd" is the universally accepted convention.

````python
import pandas as pd

print(pd.__version__)
````

# ⚠️ Common Beginner Mistakes

### 1. Forgetting to import Pandas

```python
df = pd.read_csv("data.csv")
```

**Error**

```
NameError: name 'pd' is not defined
```

**Reason**

The Pandas library was never imported.

**Solution**

```python
import pandas as pd
```

---

### 2. Importing Pandas without an alias

```python
import pandas

pandas.read_csv("data.csv")
```

Although this works, the standard convention in the Python and Data Science community is:

```python
import pandas as pd
```

Using the standard alias improves readability and collaboration.

---

### 3. Assuming Pandas is pre-installed

Some Python installations do not include Pandas.

Trying to import it may produce:

```python
ModuleNotFoundError: No module named 'pandas'
```

Install it using:

```bash
pip install pandas
```

or

```bash
conda install pandas
```
