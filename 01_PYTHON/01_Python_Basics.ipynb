## 1. Variables

### What is a Variable?

A variable is a named location in memory used to store data. It allows us to store, update, and reuse values throughout a Python program.

Python is a **dynamically typed** language, meaning you don't need to specify the data type while creating a variable. The data type is automatically determined based on the assigned value.

````python
# Creating variables

name = "Chandan"
age = 22
height = 5.11
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
````

### Key Points

- Variables store data in memory.
- Python automatically identifies the data type.
- Variable names should be meaningful.
- Variable names are case-sensitive.

### Interview Question

**Q1. Why is Python called a dynamically typed language?**

**Answer:**

Python automatically determines the data type of a variable during runtime. Therefore, the programmer does not need to declare the data type explicitly.

## 2. Data Types

### What are Data Types?

A data type defines the kind of value a variable can store. Python provides several built-in data types to represent different kinds of data.

````python
a = 10
b = 10.5
c = "Python"
d = True
e = [1, 2, 3]
f = (1, 2, 3)
g = {1, 2, 3}
h = {"name": "Chandan"}

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
````

## 3. Type Casting

### What is Type Casting?

Type casting is the process of converting a value from one data type to another.

Python provides built-in functions to perform type conversion.

Type casting can be:
- Implicit Type Casting (done automatically by Python)
- Explicit Type Casting (done manually by the programmer)

````python
# Implicit Type Casting

a = 10      # int
b = 5.5     # float

result = a + b

print(result)
print(type(result))
````

````python
# Explicit Type Casting

num = "100"

print(type(num))

num = int(num)

print(type(num))
print(num)

price = 250

print(float(price))

marks = 89.75

print(int(marks))

age = 21

print(str(age))

print(bool(1))
print(bool(0))
````

### Common Type Casting Functions

| Function | Description |
|----------|-------------|
| int() | Converts a value to an integer |
| float() | Converts a value to a floating-point number |
| str() | Converts a value to a string |
| bool() | Converts a value to a Boolean |
| list() | Converts an iterable to a list |
| tuple() | Converts an iterable to a tuple |
| set() | Converts an iterable to a set |

### Key Points

- Implicit type casting is performed automatically by Python.
- Explicit type casting is performed using built-in conversion functions.
- Some conversions are not possible and may raise a `ValueError`.

````python
# Invalid Type Casting

value = "Hello"

try:
    print(int(value))
except ValueError as e:
    print(e)
````

### Interview Question

**Q1. What is the difference between Implicit and Explicit Type Casting?**

**Answer**

| Implicit Type Casting | Explicit Type Casting |
|----------------------|-----------------------|
| Done automatically by Python | Done manually by the programmer |
| No function call required | Uses functions like int(), float(), str() |
| Safe conversion | Programmer controls the conversion |

## 4. Operators

### What are Operators?

Operators are special symbols that perform operations on one or more operands (values or variables).

Python provides different types of operators for performing arithmetic calculations, comparisons, logical operations, assignments, identity checks, membership tests, and bitwise operations.

### 4.1 Arithmetic Operators

Arithmetic operators are used to perform mathematical calculations.

````python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
````

### Arithmetic Operators

| Operator | Description | Example |
|----------|-------------|---------|
| + | Addition | a + b |
| - | Subtraction | a - b |
| * | Multiplication | a * b |
| / | Division | a / b |
| // | Floor Division | a // b |
| % | Modulus (Remainder) | a % b |
| ** | Exponent | a ** b |

### 4.2 Comparison Operators / relational operators

Comparison operators compare two values and always return either **True** or **False**.

````python
x = 20
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
````

### Comparison Operators

| Operator | Description |
|----------|-------------|
| == | Equal to |
| != | Not equal to |
| > | Greater than |
| < | Less than |
| >= | Greater than or equal to |
| <= | Less than or equal to |

### 4.3 Assignment Operators

Assignment operators are used to assign values to variables and update them.

````python
number = 10

number += 5
print(number)

number -= 2
print(number)

number *= 3
print(number)

number /= 2
print(number)

number //= 2
print(number)

number %= 3
print(number)

number **= 2
print(number)
````

### 4.4 Logical Operators

Logical operators are used to combine multiple conditions.

````python
age = 22
has_id = True

print(age >= 18 and has_id)

print(age >= 18 or False)

print(not has_id)
````

### Logical Operators

| Operator | Description |
|----------|-------------|
| and | Returns True if both conditions are True |
| or | Returns True if at least one condition is True |
| not | Reverses the Boolean value |

### 4.5 Membership Operators

Membership operators check whether a value exists in a sequence such as a list, tuple, string, or set.

````python
fruits = ["Apple", "Mango", "Orange"]

print("Apple" in fruits)

print("Banana" in fruits)

print("Banana" not in fruits)
````

### 4.6 Identity Operators

Identity operators check whether two variables refer to the same object in memory.

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)

print(a is c)

print(a == c)

print(a is not c)
````

### 4.7 Bitwise Operators

Bitwise operators perform operations on the binary representation of integers.

These operators are mainly used in low-level programming, optimization, and hardware-related applications.

````python
a = 5
b = 3

print(a & b)

print(a | b)

print(a ^ b)

print(~a)

print(a << 1)

print(a >> 1)
````

### Key Points

- Arithmetic operators perform mathematical calculations.
- Comparison operators return Boolean values.
- Assignment operators simplify variable updates.
- Logical operators combine multiple conditions.
- Membership operators check whether an element exists in a collection.
- Identity operators compare object identities, not values.
- Bitwise operators work on binary representations of numbers.

### Interview Questions

### Q1. What is the difference between `==` and `is`?

**Answer**

- `==` compares the values of two objects.
- `is` compares whether two variables refer to the same object in memory.

---

### Q2. What is the difference between `/` and `//`?

**Answer**

- `/` performs normal division and returns a float.
- `//` performs floor division and returns the quotient after removing the decimal part.

---

### Q3. Which operator is used to calculate the remainder?

**Answer**

The modulus operator `%`.

# 5. Strings

## What is a String?

A string is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Strings are one of the most commonly used data types in Python for storing and manipulating textual data.

**Note:** Strings are **immutable**, which means once a string is created, its contents cannot be modified.

````python
# Creating Strings

name = "Chandan"
city = 'Bengaluru'
message = """Welcome to Data Science"""

print(name)
print(city)
print(message)
````

## Indexing

Each character in a string has a unique position called an index.

- Positive indexing starts from **0**.
- Negative indexing starts from **-1** from the end of the string.

````python
language = "Python"

print(language[0])
print(language[2])
print(language[-1])
print(language[-3])
````

## Slicing

Slicing is used to extract a portion of a string.

**Syntax**

string[start : stop : step]

````python
language = "Python Programming"

print(language[0:6])
print(language[7:])
print(language[:6])
print(language[::2])
print(language[::-1])
````

## String Immutability

Strings are immutable, meaning their characters cannot be changed after creation.

````python
text = "Python"

try:
    text[0] = "J"
except TypeError as e:
    print(e)
````

## Common String Methods

Python provides several built-in methods to manipulate strings.

````python
text = "python programming"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
print(text.strip())
print(text.replace("python", "Java"))
print(text.split())
print("-".join(["Data", "Science"]))
print(text.startswith("python"))
print(text.endswith("ing"))
print(text.find("programming"))
print(text.count("m"))
````

## String Formatting

````python
name = "Chandan"
role = "Data Scientist"

print(f"My name is {name} and I want to become a {role}.")
````

## Key Points

- Strings are immutable.
- Indexing starts from 0.
- Negative indexing starts from -1.
- Slicing follows start:stop:step.
- Strings provide many useful built-in methods.
- f-strings are the preferred way to format strings in modern Python.

## Data Science Tip

Text data is everywhere in Data Science—customer reviews, tweets, emails, chat messages, and product descriptions.

Many NLP (Natural Language Processing) tasks start by cleaning and processing strings using methods like `lower()`, `replace()`, `split()`, and `strip()`.

# 6. Lists

## What is a List?

A List is an ordered, mutable (modifiable) collection in Python that can store elements of different data types.

Lists are created using square brackets `[]` and allow duplicate values.

Lists are one of the most frequently used data structures in Python because they are flexible and easy to manipulate.

````python
# Creating Lists

numbers = [10, 20, 30, 40]

fruits = ["Apple", "Mango", "Orange"]

mixed = [10, "Python", 3.14, True]

print(numbers)
print(fruits)
print(mixed)
````

## Characteristics of Lists

- Ordered
- Mutable
- Allows duplicate values
- Can store different data types
- Supports indexing and slicing

## Indexing

````python
fruits = ["Apple", "Mango", "Orange", "Banana"]

print(fruits[0])

print(fruits[2])

print(fruits[-1])

print(fruits[-2])
````

## Slicing

Syntax:

list[start : stop : step]

````python
numbers = [10,20,30,40,50,60,70]

print(numbers[1:5])

print(numbers[:4])

print(numbers[3:])

print(numbers[::2])

print(numbers[::-1])
````

## Updating List Elements

Lists are mutable, meaning their elements can be modified after creation.

````python
fruits = ["Apple","Mango","Orange"]

fruits[1] = "Banana"

print(fruits)
````

## Adding Elements

````python
numbers = [10,20,30]

numbers.append(40)

print(numbers)

numbers.extend([50,60])

print(numbers)

numbers.insert(2,25)

print(numbers)
````

## Removing Elements

````python
numbers = [10,20,30,40,50]

numbers.remove(30)

print(numbers)

numbers.pop()

print(numbers)

numbers.pop(1)

print(numbers)

del numbers[0]

print(numbers)
````

## Useful List Methods

````python
numbers = [50,20,10,40,30]

print(len(numbers))

print(max(numbers))

print(min(numbers))

print(sum(numbers))

numbers.sort()

print(numbers)

numbers.reverse()

print(numbers)

print(numbers.count(20))

print(numbers.index(40))
````

## Copying Lists

````python
list1 = [10,20,30]

list2 = list1.copy()

list2.append(40)

print(list1)

print(list2)
````

## Nested Lists

````python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matrix)

print(matrix[1])

print(matrix[2][1])
````

## Key Points

- Lists are mutable.
- Lists maintain insertion order.
- Lists allow duplicate values.
- Lists support indexing and slicing.
- Lists can contain multiple data types.

## Data Science Tip

Lists are useful for storing small collections of data.

However, for numerical computations involving thousands or millions of values, **NumPy arrays** are preferred because they are faster and more memory-efficient.

We'll explore NumPy in the next module.

## ⚠️ Common Mistakes & Best Practices

Understanding these common mistakes will help you write better Python code and avoid bugs during interviews and real-world projects.

## Mistake 1: Assignment vs Copy

````python
list1 = [10, 20, 30]

list2 = list1

list2.append(40)

print("List 1:", list1)
print("List 2:", list2)
````

### Why did this happen?

Using the assignment operator (`=`) does **not** create a new list.

Instead, both variables point to the **same object in memory**.

Any modification through one variable is reflected in the other.

````python
list1 = [10, 20, 30]

list2 = list1.copy()

list2.append(40)

print("List 1:", list1)
print("List 2:", list2)
````

## Mistake 2: append() vs extend()

````python
numbers = [1, 2, 3]

numbers.append([4, 5])

print(numbers)
````

````python
numbers = [1, 2, 3]

numbers.extend([4, 5])

print(numbers)
````

## Mistake 3: remove() vs pop()

````python
numbers = [10,20,30]

value = numbers.pop()

print(value)

print(numbers)
````

`pop()` returns the removed element.

`remove()` does **not** return the removed element.

## Mistake 4: sort() returns None

````python
numbers = [30,10,20]

result = numbers.sort()

print(result)
````

````python
numbers = [30,10,20]

numbers.sort()

print(numbers)
````

## Mistake 5: Removing a value that doesn't exist

````python
# numbers = [10,20,30]

# numbers.remove(100)
''' ValueError: list.remove(x): x not in list'''
````

````python
numbers = [10,20,30]

if 100 in numbers:
    numbers.remove(100)
````

## Best Practices

- Use meaningful variable names.
- Use `copy()` when creating a separate list.
- Use `append()` for a single element.
- Use `extend()` for multiple elements.
- Prefer `sorted()` if you need to preserve the original list.
- Check whether a value exists before using `remove()`.

## Bonus Interview Questions

### Q1. What is the difference between shallow copy and deep copy?

> A shallow copy copies only the outer object. Nested objects are still shared.

> A deep copy creates completely independent copies of all nested objects.

---

### Q2. Why is `sort()` faster than `sorted()`?

`sort()` modifies the existing list in place, so it uses less memory.

`sorted()` creates a new list, which requires additional memory.

---

### Q3. Why is `append()` O(1) while `insert(0, value)` is O(n)?

`append()` adds an element at the end of the list.

`insert(0, value)` shifts every existing element one position to the right, making it a linear-time operation.

Neither List nor Tuple.

# 7. Tuples

## What is a Tuple?

A Tuple is an **ordered**, **immutable** collection in Python that can store elements of different data types.

Tuples are created using parentheses `()` and allow duplicate values.

Since tuples are immutable, their elements cannot be modified after creation. This makes them faster and safer than lists for storing fixed data.

````python
# Creating Tuples

numbers = (10, 20, 30, 40)

fruits = ("Apple", "Mango", "Orange")

mixed = (10, "Python", 3.14, True)

print(numbers)
print(fruits)
print(mixed)
````

## Characteristics of Tuples

- Ordered
- Immutable
- Allows duplicate values
- Can store different data types
- Supports indexing and slicing
- Faster than lists for read-only data

## Indexing

````python
fruits = ("Apple", "Mango", "Orange", "Banana")

print(fruits[0])
print(fruits[2])
print(fruits[-1])
print(fruits[-2])
````

## Slicing

Syntax:

tuple[start : stop : step]

````python
numbers = (10, 20, 30, 40, 50, 60, 70)

print(numbers[1:5])
print(numbers[:4])
print(numbers[3:])
print(numbers[::2])
print(numbers[::-1])
````

## Tuple Packing and Unpacking

Packing means storing multiple values into a single tuple.

Unpacking means extracting tuple elements into individual variables.

````python
# Packing

student = ("Chandan", 22, "Data Science")

print(student)

# Unpacking

name, age, course = student

print(name)
print(age)
print(course)
````

## Important Tuple Methods

Tuples have only two built-in methods because they are immutable.

````python
numbers = (10, 20, 30, 20, 40, 20)

print(numbers.count(20))

print(numbers.index(30))
````

## Tuple Immutability

Since tuples are immutable, elements cannot be modified after creation.

````python
numbers = (10, 20, 30)

try:
    numbers[0] = 100
except TypeError as e:
    print(e)
````

## Tuple vs List

| Feature | List | Tuple |
|----------|------|-------|
| Syntax | [] | () |
| Mutable | ✅ Yes | ❌ No |
| Ordered | ✅ Yes | ✅ Yes |
| Allows Duplicates | ✅ Yes | ✅ Yes |
| Performance | Slightly Slower | Slightly Faster |
| Methods | Many | Only 2 |

## Key Points

- Tuples are immutable.
- Tuples maintain insertion order.
- Tuples allow duplicate values.
- Tuples support indexing and slicing.
- Tuples consume less memory than lists.

## Data Science Tip

Tuples are useful for storing fixed information such as dataset dimensions, RGB color values, coordinates, or model configuration values that should not change during program execution.

⚠️ Common Mistakes & Best Practices

````python
numbers = (10)

print(type(numbers))
````

### Why did this happen?

Python treats `(10)` as an integer enclosed in parentheses, **not** as a tuple.

To create a tuple with a single element, you must include a trailing comma.

````python
numbers = (10,)

print(type(numbers))
````

````python
person = ("Chandan", ["Python", "SQL"])

person[1].append("Machine Learning")

print(person)
````

### Important Note

Although tuples are immutable, they can contain mutable objects such as lists.

The tuple itself cannot be modified, but the mutable object inside it can.

# 8. Sets

## What is a Set?

A Set is an **unordered**, **mutable** collection in Python that stores **unique elements**.

Sets are created using curly braces `{}` or the `set()` constructor.

Unlike lists and tuples, sets do not allow duplicate values and do not support indexing.

````python
# Creating Sets

numbers = {10, 20, 30, 40}

fruits = {"Apple", "Mango", "Orange"}

mixed = {10, "Python", 3.14, True}

print(numbers)
print(fruits)
print(mixed)
````

## Characteristics of Sets

- Unordered
- Mutable
- Does not allow duplicate values
- Supports mathematical set operations
- Does not support indexing or slicing
- Fast membership testing

## Duplicate Values

One of the biggest advantages of a Set is that it automatically removes duplicate values.

````python
numbers = {10, 20, 20, 30, 40, 40, 50}

print(numbers)
````

### Visual Representation

```text
Input

{10,20,20,30,40,40}

        ↓

Output

{10,20,30,40}
```

## Creating an Empty Set

Use `set()` to create an empty set.

**Note:** `{}` creates an empty dictionary, not a set.

````python
empty_set = set()

print(type(empty_set))

empty_dict = {}

print(type(empty_dict))
````

## Adding Elements

````python
numbers = {10, 20, 30}

numbers.add(40)

print(numbers)

numbers.update([50, 60])

print(numbers)
````

## Removing Elements

````python
numbers = {10, 20, 30, 40}

numbers.remove(20)

print(numbers)

numbers.discard(30)

print(numbers)

removed = numbers.pop()

print("Removed:", removed)

print(numbers)
````

## Set Operations

Python supports mathematical operations on sets.

````python
set1 = {1, 2, 3, 4}

set2 = {3, 4, 5, 6}

print("Union:", set1 | set2)

print("Intersection:", set1 & set2)

print("Difference:", set1 - set2)

print("Symmetric Difference:", set1 ^ set2)
````

## Membership Testing

Checking whether an element exists in a Set is very fast.

````python
fruits = {"Apple", "Mango", "Orange"}

print("Apple" in fruits)

print("Banana" in fruits)
````

## Set vs List

| Feature | List | Set |
|----------|------|-----|
| Ordered | ✅ | ❌ |
| Duplicate Values | ✅ | ❌ |
| Mutable | ✅ | ✅ |
| Indexing | ✅ | ❌ |
| Membership Search | O(n) | O(1) Average |

## Key Points

- Sets store only unique elements.
- Sets are unordered.
- Sets are mutable.
- Sets do not support indexing.
- Membership checking is very fast.

## Data Science Tip

Sets are commonly used for:

- Removing duplicate records
- Finding common values between datasets
- Comparing categories
- Fast membership testing during data preprocessing

⚠️ Common Mistakes & Best Practices

Mistake 1: Creating an Empty Set

````python
empty = {}

print(type(empty))
````

````python
# correct way to create an empty list

empty = set()

print(type(empty))
````

Mistake 2: Accessing by Index

````python
''' numbers = {10, 20, 30}

print(numbers[0])

TypeError: 'set' object is not subscriptable
'''
````

This raises a `TypeError` because Sets are unordered and do not support indexing.

Mistake 3: remove() vs discard()

````python
numbers = {10, 20, 30}

numbers.discard(100)

print(numbers)
````

### Difference

- `remove()` raises a `KeyError` if the element is not found.
- `discard()` does nothing if the element is not found.

In most real-world applications, `discard()` is safer when you're unsure whether the element exists.

| Allows duplicates | Unique elements only |
| Supports indexing | No indexing |
| Membership O(n) | Membership O(1) |

---

### Q3. Why are Sets faster than Lists for membership checking?

Sets use a **hash table** internally.

Instead of checking elements one by one, Python computes a hash and directly looks up the element, making average lookup **O(1)**.

---

### Q4. Difference between remove() and discard()?

- `remove()` raises an error if the element is absent.
- `discard()` silently ignores missing elements.

---

## Scenario-Based

### Q5. You have a list containing one million customer IDs with many duplicates.

How would you quickly remove duplicates?

**Answer:**

Convert the list into a Set.

```python
unique_ids = set(customer_ids)
```

---

### Q6. Can a Set contain a List?

No.

Lists are mutable and therefore unhashable.

A Set can only contain hashable (immutable) elements such as integers, strings, floats, and tuples (if all their elements are hashable).

## What is a Dictionary?

A Dictionary is an **unordered** (insertion order is preserved from Python 3.7+), **mutable** collection that stores data as **key-value pairs**.

Each key in a dictionary must be **unique** and **hashable**, while values can be of any data type.

Dictionaries are created using curly braces `{}`.

````python
# Creating Dictionaries

student = {
    "name": "Chandan",
    "age": 22,
    "course": "Data Science"
}

print(student)
````

## Characteristics of Dictionaries

- Stores data as key-value pairs
- Mutable
- Keys must be unique
- Keys must be hashable
- Values can be duplicated
- Fast lookup using keys

## Accessing Values

````python
student = {
    "name": "Chandan",
    "age": 22,
    "course": "Data Science"
}

print(student["name"])

print(student["course"])

print(student.get("age"))
````

## Adding and Updating Elements

````python
student = {
    "name": "Chandan",
    "age": 22
}

student["course"] = "Data Science"

student["age"] = 23

print(student)
````

## Removing Elements

````python
student = {
    "name": "Chandan",
    "age": 22,
    "course": "Data Science"
}

removed = student.pop("age")

print("Removed:", removed)

print(student)

student.clear()

print(student)
````

## Important Dictionary Methods

````python
student = {
    "name": "Chandan",
    "age": 22,
    "course": "Data Science"
}

print(student.keys())

print(student.values())

print(student.items())

print(student.get("name"))

print(student.get("salary", "Not Available"))
````

## Iterating Through a Dictionary

````python
student = {
    "name": "Chandan",
    "age": 22,
    "course": "Data Science"
}

for key, value in student.items():
    print(key, ":", value)
````

````python
student = {
    "name": "Chandan",
    "name": "Rahul"
}

print(student)
````

Mistake 3: Using a List as a Key

````python
''' 
data = {
    [1, 2]: "Numbers"
}
TypeError: unhashable type: 'list'
'''
````

Lists are mutable and therefore cannot be used as dictionary keys.

## Behind the Scenes: Hashing

Python dictionaries use a **hash table** internally.

When you access a value using a key:

```python
student["name"]
```

Python computes a **hash value** for the key `"name"` and directly jumps to the location where the value is stored.

Because it doesn't search one element at a time, dictionary lookup is **O(1)** on average.

This is why dictionaries are much faster than searching through a list.

   │

 │             │
 ▼             ▼

# 12. List Comprehension

## What is List Comprehension?

List Comprehension is a concise and Pythonic way to create a new list from an existing iterable such as a list, tuple, string, or range.

It combines a loop and optional conditions into a single line of code, making programs shorter, more readable, and often faster than traditional loops.

## Why Use List Comprehension?

Instead of writing multiple lines using a `for` loop, list comprehension allows us to create a list in a single line.

It improves:
- Readability
- Performance
- Code simplicity

## Syntax

```python
new_list = [expression for item in iterable]
```

Where:

- **expression** → Value to be added to the list.
- **item** → Current element from the iterable.
- **iterable** → List, tuple, string, range, etc.

````python
numbers = [1, 2, 3, 4, 5]

square = [num ** 2 for num in numbers]

print(square)
````

## Traditional Loop vs List Comprehension

````python
numbers = [1, 2, 3, 4, 5]

square = []

for num in numbers:
    square.append(num ** 2)

print(square)
````

````python
numbers = [1, 2, 3, 4, 5]

square = [num ** 2 for num in numbers]

print(square)
````

## Example 1: Square of Numbers

````python
square = [x ** 2 for x in range(1, 6)]

print(square)
````

## Example 2: Even Numbers

````python
even = [x for x in range(1, 21) if x % 2 == 0]

print(even)
````

## Example 3: Odd Numbers

````python
odd = [x for x in range(1, 21) if x % 2 != 0]

print(odd)
````

## Example 4: Convert Strings to Uppercase

````python
fruits = ["apple", "banana", "mango"]

upper = [fruit.upper() for fruit in fruits]

print(upper)
````

## Example 5: Length of Each Word

````python
languages = ["Python", "Java", "C++", "JavaScript"]

length = [len(lang) for lang in languages]

print(length)
````

## Example 6: Using if Condition

````python
numbers = [10, 15, 20, 25, 30]

result = [num for num in numbers if num > 20]

print(result)
````

## Example 7: Using if-else

````python
numbers = [2, 5, 8, 11, 14]

result = ["Even" if num % 2 == 0 else "Odd" for num in numbers]

print(result)
````

## Example 8: Nested List Comprehension

````python
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]

print(matrix)
````

## Key Points

- Creates a new list.
- Works with any iterable.
- Supports conditions.
- More readable than simple loops.
- Often faster than traditional loops.

## Data Science Tip

List comprehensions are frequently used in Data Science for:

- Data cleaning
- Feature transformation
- Filtering records
- Creating new features
- Preprocessing datasets

However, when working with large datasets, **NumPy** and **Pandas** vectorized operations are usually preferred because they are much faster.
