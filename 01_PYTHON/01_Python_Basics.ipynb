## 1. Variables

### What is a Variable?

A variable is a named location in memory used to store data. It allows us to store, update, and reuse values throughout a Python program.

Python is a **dynamically typed** language, meaning you don't need to specify the data type while creating a variable. The data type is automatically determined based on the assigned value.

````python
# Creating variables

name = "Chandan"
age = 22
height = 5.11
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
````

### Key Points

- Variables store data in memory.
- Python automatically identifies the data type.
- Variable names should be meaningful.
- Variable names are case-sensitive.

### Interview Question

**Q1. Why is Python called a dynamically typed language?**

**Answer:**

Python automatically determines the data type of a variable during runtime. Therefore, the programmer does not need to declare the data type explicitly.

## 2. Data Types

### What are Data Types?

A data type defines the kind of value a variable can store. Python provides several built-in data types to represent different kinds of data.

````python
a = 10
b = 10.5
c = "Python"
d = True
e = [1, 2, 3]
f = (1, 2, 3)
g = {1, 2, 3}
h = {"name": "Chandan"}

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
````

### Common Python Data Types

| Data Type | Example |
|------------|----------|
| int | 10 |
| float | 10.5 |
| str | "Python" |
| bool | True |
| list | [1,2,3] |
| tuple | (1,2,3) |
| set | {1,2,3} |
| dict | {"name":"Chandan"} |

### Interview Question

**Q2. What is the difference between mutable and immutable data types?**

**Answer:**

- Mutable objects can be modified after creation (List, Dictionary, Set).
- Immutable objects cannot be modified after creation (String, Tuple, Integer, Float).

## 3. Type Casting

### What is Type Casting?

Type casting is the process of converting a value from one data type to another.

Python provides built-in functions to perform type conversion.

Type casting can be:
- Implicit Type Casting (done automatically by Python)
- Explicit Type Casting (done manually by the programmer)

````python
# Implicit Type Casting

a = 10      # int
b = 5.5     # float

result = a + b

print(result)
print(type(result))
````

````python
# Explicit Type Casting

num = "100"

print(type(num))

num = int(num)

print(type(num))
print(num)

price = 250

print(float(price))

marks = 89.75

print(int(marks))

age = 21

print(str(age))

print(bool(1))
print(bool(0))
````

### Common Type Casting Functions

| Function | Description |
|----------|-------------|
| int() | Converts a value to an integer |
| float() | Converts a value to a floating-point number |
| str() | Converts a value to a string |
| bool() | Converts a value to a Boolean |
| list() | Converts an iterable to a list |
| tuple() | Converts an iterable to a tuple |
| set() | Converts an iterable to a set |

### Key Points

- Implicit type casting is performed automatically by Python.
- Explicit type casting is performed using built-in conversion functions.
- Some conversions are not possible and may raise a `ValueError`.

````python
# Invalid Type Casting

value = "Hello"

try:
    print(int(value))
except ValueError as e:
    print(e)
````

### Interview Question

**Q1. What is the difference between Implicit and Explicit Type Casting?**

**Answer**

| Implicit Type Casting | Explicit Type Casting |
|----------------------|-----------------------|
| Done automatically by Python | Done manually by the programmer |
| No function call required | Uses functions like int(), float(), str() |
| Safe conversion | Programmer controls the conversion |

## 4. Operators

### What are Operators?

Operators are special symbols that perform operations on one or more operands (values or variables).

Python provides different types of operators for performing arithmetic calculations, comparisons, logical operations, assignments, identity checks, membership tests, and bitwise operations.

### 4.1 Arithmetic Operators

Arithmetic operators are used to perform mathematical calculations.

````python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
````

### Arithmetic Operators

| Operator | Description | Example |
|----------|-------------|---------|
| + | Addition | a + b |
| - | Subtraction | a - b |
| * | Multiplication | a * b |
| / | Division | a / b |
| // | Floor Division | a // b |
| % | Modulus (Remainder) | a % b |
| ** | Exponent | a ** b |

### 4.2 Comparison Operators / relational operators

Comparison operators compare two values and always return either **True** or **False**.

````python
x = 20
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
````

### Comparison Operators

| Operator | Description |
|----------|-------------|
| == | Equal to |
| != | Not equal to |
| > | Greater than |
| < | Less than |
| >= | Greater than or equal to |
| <= | Less than or equal to |

### 4.3 Assignment Operators

Assignment operators are used to assign values to variables and update them.

````python
number = 10

number += 5
print(number)

number -= 2
print(number)

number *= 3
print(number)

number /= 2
print(number)

number //= 2
print(number)

number %= 3
print(number)

number **= 2
print(number)
````

### 4.4 Logical Operators

Logical operators are used to combine multiple conditions.

````python
age = 22
has_id = True

print(age >= 18 and has_id)

print(age >= 18 or False)

print(not has_id)
````

### Logical Operators

| Operator | Description |
|----------|-------------|
| and | Returns True if both conditions are True |
| or | Returns True if at least one condition is True |
| not | Reverses the Boolean value |

### 4.5 Membership Operators

Membership operators check whether a value exists in a sequence such as a list, tuple, string, or set.

````python
fruits = ["Apple", "Mango", "Orange"]

print("Apple" in fruits)

print("Banana" in fruits)

print("Banana" not in fruits)
````

### 4.6 Identity Operators

Identity operators check whether two variables refer to the same object in memory.

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)

print(a is c)

print(a == c)

print(a is not c)
````

### 4.7 Bitwise Operators

Bitwise operators perform operations on the binary representation of integers.

These operators are mainly used in low-level programming, optimization, and hardware-related applications.

````python
a = 5
b = 3

print(a & b)

print(a | b)

print(a ^ b)

print(~a)

print(a << 1)

print(a >> 1)
````

### Key Points

- Arithmetic operators perform mathematical calculations.
- Comparison operators return Boolean values.
- Assignment operators simplify variable updates.
- Logical operators combine multiple conditions.
- Membership operators check whether an element exists in a collection.
- Identity operators compare object identities, not values.
- Bitwise operators work on binary representations of numbers.

### Interview Questions

### Q1. What is the difference between `==` and `is`?

**Answer**

- `==` compares the values of two objects.
- `is` compares whether two variables refer to the same object in memory.

---

### Q2. What is the difference between `/` and `//`?

**Answer**

- `/` performs normal division and returns a float.
- `//` performs floor division and returns the quotient after removing the decimal part.

---

### Q3. Which operator is used to calculate the remainder?

**Answer**

The modulus operator `%`.

# 5. Strings

## What is a String?

A string is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Strings are one of the most commonly used data types in Python for storing and manipulating textual data.

**Note:** Strings are **immutable**, which means once a string is created, its contents cannot be modified.

````python
# Creating Strings

name = "Chandan"
city = 'Bengaluru'
message = """Welcome to Data Science"""

print(name)
print(city)
print(message)
````

## Indexing

Each character in a string has a unique position called an index.

- Positive indexing starts from **0**.
- Negative indexing starts from **-1** from the end of the string.

````python
language = "Python"

print(language[0])
print(language[2])
print(language[-1])
print(language[-3])
````

## Slicing

Slicing is used to extract a portion of a string.

**Syntax**

string[start : stop : step]

````python
language = "Python Programming"

print(language[0:6])
print(language[7:])
print(language[:6])
print(language[::2])
print(language[::-1])
````

## String Immutability

Strings are immutable, meaning their characters cannot be changed after creation.

````python
text = "Python"

try:
    text[0] = "J"
except TypeError as e:
    print(e)
````

## Common String Methods

Python provides several built-in methods to manipulate strings.

````python
text = "python programming"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
print(text.strip())
print(text.replace("python", "Java"))
print(text.split())
print("-".join(["Data", "Science"]))
print(text.startswith("python"))
print(text.endswith("ing"))
print(text.find("programming"))
print(text.count("m"))
````

## String Formatting

````python
name = "Chandan"
role = "Data Scientist"

print(f"My name is {name} and I want to become a {role}.")
````

## Key Points

- Strings are immutable.
- Indexing starts from 0.
- Negative indexing starts from -1.
- Slicing follows start:stop:step.
- Strings provide many useful built-in methods.
- f-strings are the preferred way to format strings in modern Python.

## Data Science Tip

Text data is everywhere in Data Science—customer reviews, tweets, emails, chat messages, and product descriptions.

Many NLP (Natural Language Processing) tasks start by cleaning and processing strings using methods like `lower()`, `replace()`, `split()`, and `strip()`.

# 6. Lists

## What is a List?

A List is an ordered, mutable (modifiable) collection in Python that can store elements of different data types.

Lists are created using square brackets `[]` and allow duplicate values.

Lists are one of the most frequently used data structures in Python because they are flexible and easy to manipulate.

````python
# Creating Lists

numbers = [10, 20, 30, 40]

fruits = ["Apple", "Mango", "Orange"]

mixed = [10, "Python", 3.14, True]

print(numbers)
print(fruits)
print(mixed)
````

## Characteristics of Lists

- Ordered
- Mutable
- Allows duplicate values
- Can store different data types
- Supports indexing and slicing

## Indexing

````python
fruits = ["Apple", "Mango", "Orange", "Banana"]

print(fruits[0])

print(fruits[2])

print(fruits[-1])

print(fruits[-2])
````

## Slicing

Syntax:

list[start : stop : step]

````python
numbers = [10,20,30,40,50,60,70]

print(numbers[1:5])

print(numbers[:4])

print(numbers[3:])

print(numbers[::2])

print(numbers[::-1])
````

## Updating List Elements

Lists are mutable, meaning their elements can be modified after creation.

````python
fruits = ["Apple","Mango","Orange"]

fruits[1] = "Banana"

print(fruits)
````

## Adding Elements

````python
numbers = [10,20,30]

numbers.append(40)

print(numbers)

numbers.extend([50,60])

print(numbers)

numbers.insert(2,25)

print(numbers)
````

## Removing Elements

````python
numbers = [10,20,30,40,50]

numbers.remove(30)

print(numbers)

numbers.pop()

print(numbers)

numbers.pop(1)

print(numbers)

del numbers[0]

print(numbers)
````

## Useful List Methods

````python
numbers = [50,20,10,40,30]

print(len(numbers))

print(max(numbers))

print(min(numbers))

print(sum(numbers))

numbers.sort()

print(numbers)

numbers.reverse()

print(numbers)

print(numbers.count(20))

print(numbers.index(40))
````

## Copying Lists

````python
list1 = [10,20,30]

list2 = list1.copy()

list2.append(40)

print(list1)

print(list2)
````

## Nested Lists

````python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

print(matrix)

print(matrix[1])

print(matrix[2][1])
````

## Key Points

- Lists are mutable.
- Lists maintain insertion order.
- Lists allow duplicate values.
- Lists support indexing and slicing.
- Lists can contain multiple data types.

## Data Science Tip

Lists are useful for storing small collections of data.

However, for numerical computations involving thousands or millions of values, **NumPy arrays** are preferred because they are faster and more memory-efficient.

We'll explore NumPy in the next module.

## ⚠️ Common Mistakes & Best Practices

Understanding these common mistakes will help you write better Python code and avoid bugs during interviews and real-world projects.

## Mistake 1: Assignment vs Copy

````python
list1 = [10, 20, 30]

list2 = list1

list2.append(40)

print("List 1:", list1)
print("List 2:", list2)
````

### Why did this happen?

Using the assignment operator (`=`) does **not** create a new list.

Instead, both variables point to the **same object in memory**.

Any modification through one variable is reflected in the other.

````python
list1 = [10, 20, 30]

list2 = list1.copy()

list2.append(40)

print("List 1:", list1)
print("List 2:", list2)
````

## Mistake 2: append() vs extend()

````python
numbers = [1, 2, 3]

numbers.append([4, 5])

print(numbers)
````

````python
numbers = [1, 2, 3]

numbers.extend([4, 5])

print(numbers)
````

## Mistake 3: remove() vs pop()

````python
numbers = [10,20,30]

value = numbers.pop()

print(value)

print(numbers)
````

`pop()` returns the removed element.

`remove()` does **not** return the removed element.

## Mistake 4: sort() returns None

````python
numbers = [30,10,20]

result = numbers.sort()

print(result)
````

````python
numbers = [30,10,20]

numbers.sort()

print(numbers)
````

## Mistake 5: Removing a value that doesn't exist

````python
# numbers = [10,20,30]

# numbers.remove(100)
''' ValueError: list.remove(x): x not in list'''
````

````python
numbers = [10,20,30]

if 100 in numbers:
    numbers.remove(100)
````

## Best Practices

- Use meaningful variable names.
- Use `copy()` when creating a separate list.
- Use `append()` for a single element.
- Use `extend()` for multiple elements.
- Prefer `sorted()` if you need to preserve the original list.
- Check whether a value exists before using `remove()`.

## Bonus Interview Questions

### Q1. What is the difference between shallow copy and deep copy?

> A shallow copy copies only the outer object. Nested objects are still shared.

> A deep copy creates completely independent copies of all nested objects.

---

### Q2. Why is `sort()` faster than `sorted()`?

`sort()` modifies the existing list in place, so it uses less memory.

`sorted()` creates a new list, which requires additional memory.

---

### Q3. Why is `append()` O(1) while `insert(0, value)` is O(n)?

`append()` adds an element at the end of the list.

`insert(0, value)` shifts every existing element one position to the right, making it a linear-time operation.

---

### Q4. Which data structure would you choose for storing millions of numerical values in Machine Learning?

Answer:

Neither List nor Tuple.

Use **NumPy arrays** because they are optimized for numerical computation and consume less memory.
