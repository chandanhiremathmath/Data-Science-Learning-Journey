# 2. Functions

## What is a Function?

A function is a reusable block of code that performs a specific task.

Instead of writing the same code multiple times, we write it once inside a function and call it whenever required.

Functions help make programs more organized, reusable, and easier to maintain.

## Why Do We Need Functions?

Imagine you need to calculate the average marks of 100 students.

Without functions, you would write the same logic repeatedly.

By creating a function once, you can reuse it whenever needed.

This reduces duplicate code and makes programs easier to understand.

## Advantages of Functions

- Avoids code duplication
- Improves code readability
- Makes debugging easier
- Promotes code reusability
- Makes large programs easier to manage

## Syntax of a Function

```python
def function_name(parameters):
    # function body
    return value
```

### Explanation

- `def` → Keyword used to define a function.
- `function_name` → Name of the function.
- `parameters` → Inputs accepted by the function.
- `return` → Returns the result to the caller.
```

## Types of User-Defined Functions

Based on the presence of arguments and return values, user-defined functions are classified into four types:

1. Without Arguments and Without Return Value
2. Without Arguments and With Return Value
3. With Arguments and Without Return Value
4. With Arguments and With Return Value

Understanding these four types is important because they form the foundation of function programming in Python.

## 1. Without Arguments and Without Return Value

In this type of function:

- No arguments are passed while calling the function.
- The function does not return any value.
- It simply performs a task.

````python
def greet():
    print("Welcome to Data Science")

greet()
````

### Real-World Example

A company displays a welcome message whenever the application starts.

Since no input is required and nothing needs to be returned, this type of function is suitable.

````python
def show_company_name():
    print("ABC Technologies Pvt Ltd")

show_company_name()
````

### Key Points

- No input is required.
- No value is returned.
- Mainly used for displaying messages or performing simple tasks.

## 2. Without Arguments and With Return Value

In this type:

- No arguments are passed.
- The function returns a value.

````python
def get_pi():
    return 3.14159

value = get_pi()

print(value)
````

````python
def get_current_course():
    return "Data Science"

course = get_current_course()

print(course)
````

### Real-World Example

Suppose an application stores the company name internally.

Whenever needed, a function returns the company name without taking any input.

````python
def company_name():
    return "OpenAI"

print(company_name())
````

### Key Points

- No input is required.
- The function returns a value.
- Useful when data is already available inside the function.

## 3. With Arguments and Without Return Value

In this type:

- Arguments are passed while calling the function.
- The function performs a task.
- No value is returned.

````python
def greet(name):
    print(f"Welcome {name}")

greet("Chandan")
````

````python
def student(name, course):
    print("Name :", name)
    print("Course :", course)

student("Chandan", "Data Science")
````

### Real-World Example

An online shopping application greets every customer using their name.

The customer's name is passed as an argument, and the function displays a personalized welcome message.

````python
def welcome_customer(name):
    print(f"Hello {name}, Welcome to Amazon")

welcome_customer("Rahul")
````

### Key Points

- Input is provided through arguments.
- No value is returned.
- Mostly used for displaying customized information.

## 4. With Arguments and With Return Value

This is the most commonly used type of function.

- Arguments are passed.
- The function processes the data.
- The result is returned.

````python
def add(a, b):
    return a + b

result = add(10, 20)

print(result)
````

````python
def calculate_average(marks):
    return sum(marks) / len(marks)

student_marks = [80, 90, 85, 95]

average = calculate_average(student_marks)

print(average)
````

### Real-World Example

Machine Learning models receive input data, process it, and return predictions.

Most ML functions follow this type.

````python
def calculate_salary(basic_salary, bonus):
    return basic_salary + bonus

salary = calculate_salary(50000, 8000)

print("Total Salary =", salary)
````

## Parameters vs Arguments

Many beginners get confused between parameters and arguments.

Although they look similar, they are used at different stages of a function.

### Parameters

Parameters are the variables defined in the function definition.

They act as placeholders to receive values when the function is called.

````python
def student(name, course):
    print(name)
    print(course)
````

In the above example:

- `name` and `course` are **parameters** because they are defined inside the function declaration.

### Arguments

Arguments are the actual values passed to a function during the function call.

````python
def student(name, course):
    print(name)
    print(course)

student("Chandan", "Data Science")
````

In the above example:

- `"Chandan"` and `"Data Science"` are **arguments** because they are passed while calling the function.

### Quick Comparison

| Parameters | Arguments |
|------------|-----------|
| Defined in the function declaration | Passed during the function call |
| Act as placeholders | Actual values |
| Receive data | Provide data |

````python
def add(a, b):      # Parameters
    return a + b

result = add(10, 20)   # Arguments

print(result)
````

### Interview Question

**Q. What is the difference between parameters and arguments?**

**Answer:**

Parameters are variables declared in the function definition, whereas arguments are the actual values supplied when the function is called.

# Types of Arguments

Python provides different ways to pass arguments to a function.

They are:

1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable-Length Arguments (*args)
5. Keyword Variable-Length Arguments (**kwargs)

## 1. Positional Arguments

In positional arguments, values are assigned based on their position.

The order of arguments must match the order of parameters.

````python
def student(name, course):
    print("Name :", name)
    print("Course :", course)

student("Chandan", "Data Science")
````

````python
def subtract(a, b):
    print(a - b)

subtract(20, 10)
````

### Key Points

- Order is important.
- Number of arguments should match the number of parameters.

## 2. Keyword Arguments

In keyword arguments, values are passed using the parameter names.

Since each value is associated with a parameter name, the order does not matter.

````python
def student(name, course):
    print("Name :", name)
    print("Course :", course)

student(course="Data Science", name="Chandan")
````

````python
def employee(name, department):
    print(name)
    print(department)

employee(department="HR", name="Rahul")
````

### Key Points

- Order does not matter.
- Improves code readability.
- Useful when a function has many parameters.

## 3. Default Arguments

A default argument is a parameter that already has a predefined value.

If no value is passed while calling the function, the default value is used.

````python
def greet(name="Guest"):
    print("Welcome", name)

greet()
greet("Chandan")
````

````python
def calculate_interest(amount, rate=5):
    interest = amount * rate / 100
    return interest

print(calculate_interest(10000))

print(calculate_interest(10000, 8))
````

### Key Points

- Default values make functions more flexible.
- You can override the default value by passing a new value.
