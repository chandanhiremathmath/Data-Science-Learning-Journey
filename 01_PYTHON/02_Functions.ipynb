# 2. Functions

## What is a Function?

A function is a reusable block of code that performs a specific task.

Instead of writing the same code multiple times, we write it once inside a function and call it whenever needed.

Functions make code:
- Reusable
- Easy to read
- Easy to debug
- Easy to maintain

## Why Do We Need Functions?

Imagine you need to calculate the average marks of 100 students.

Without functions, you may end up writing the same logic again and again.

With a function, you write the logic once and simply call it whenever required.

This reduces duplicate code and makes programs cleaner.

````python
# Without Function

marks = [85, 90, 78]

average = sum(marks) / len(marks)

print(average)
````

````python
# Using Function

def calculate_average(marks):
    return sum(marks) / len(marks)

student_marks = [85, 90, 78]

print(calculate_average(student_marks))
````

## Syntax

```python
def function_name(parameters):
    # function body
    return value
```

- `def` → Keyword used to define a function.
- `function_name` → Name of the function.
- `parameters` → Inputs accepted by the function.
- `return` → Sends the result back to the caller.

## Creating and Calling a Function

````python
def greet():
    print("Welcome to Data Science!")

greet()
````

## Function with Parameters

Parameters are the variables defined in the function declaration.

They allow a function to accept input values.

````python
def greet(name):
    print(f"Hello, {name}")

greet("Chandan")
greet("Rahul")
````

## Function with Multiple Parameters

````python
def add(a, b):
    print("Sum =", a + b)

add(10, 20)
````

## Parameters vs Arguments

These two terms are often confused in interviews.

- **Parameters** are the variables written while defining a function.
- **Arguments** are the actual values passed when calling the function.

````python
def add(a, b):      # a and b are parameters
    print(a + b)

add(10, 20)         # 10 and 20 are arguments
````

## Types of Arguments

Python supports different ways of passing arguments to functions:

1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable-Length Arguments (*args)
5. Keyword Variable-Length Arguments (**kwargs)

In this section, we'll learn the first three.

## Positional Arguments

The values are assigned based on their position.

````python
def student(name, course):
    print(name)
    print(course)

student("Chandan", "Data Science")
````

## Keyword Arguments

Here, we specify the parameter names while calling the function.

The order of arguments does not matter.

````python
def student(name, course):
    print(name)
    print(course)

student(course="Data Science", name="Chandan")
````

## Default Arguments

A default value is assigned to a parameter.

If no value is passed, the default value is used.

````python
def greet(name="Guest"):
    print(f"Hello, {name}")

greet()

greet("Chandan")
````

## return Statement

The `return` statement sends the result back to the caller.

Unlike `print()`, the returned value can be stored in a variable and used later.

````python
def square(number):
    return number ** 2

result = square(5)

print(result)
````

````python
def multiply(a, b):
    return a * b

value = multiply(4, 6)

print(value)
````

## print() vs return()

Many beginners think both are the same, but they serve different purposes.

````python
def using_print():
    print("Hello")

result = using_print()

print(result)
````

````python
def using_return():
    return "Hello"

result = using_return()

print(result)
````

## Key Points

- Functions avoid code duplication.
- A function can accept zero or more parameters.
- `return` sends a value back to the caller.
- Keyword arguments improve readability.
- Default arguments make functions more flexible.
