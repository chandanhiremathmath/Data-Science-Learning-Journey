## What is a Function?

A function is a reusable block of code that performs a specific task.

Instead of writing the same code multiple times, we write it once inside a function and call it whenever required.

Functions help make programs more organized, reusable, and easier to maintain.

## Why Do We Need Functions?

Imagine you need to calculate the average marks of 100 students.

Without functions, you would write the same logic repeatedly.

By creating a function once, you can reuse it whenever needed.

This reduces duplicate code and makes programs easier to understand.

## Advantages of Functions

- Avoids code duplication
- Improves code readability
- Makes debugging easier
- Promotes code reusability
- Makes large programs easier to manage

## Parameters vs Arguments

Many beginners get confused between parameters and arguments.

Although they look similar, they are used at different stages of a function.

### Parameters

Parameters are the variables defined in the function definition.

They act as placeholders to receive values when the function is called.

````python
def student(name, course):
    print(name)
    print(course)
````

In the above example:

- `name` and `course` are **parameters** because they are defined inside the function declaration.

### Arguments

Arguments are the actual values passed to a function during the function call.

````python
def student(name, course):
    print(name)
    print(course)

student("Chandan", "Data Science")
````

In the above example:

- `"Chandan"` and `"Data Science"` are **arguments** because they are passed while calling the function.

### Quick Comparison

| Parameters | Arguments |
|------------|-----------|
| Defined in the function declaration | Passed during the function call |
| Act as placeholders | Actual values |
| Receive data | Provide data |

````python
def add(a, b):      # Parameters
    return a + b

result = add(10, 20)   # Arguments

print(result)
````

### Interview Question

**Q. What is the difference between parameters and arguments?**

**Answer:**

Parameters are variables declared in the function definition, whereas arguments are the actual values supplied when the function is called.

# Types of Arguments

Python provides different ways to pass arguments to a function.

They are:

1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable-Length Arguments (*args)
5. Keyword Variable-Length Arguments (**kwargs)

## 1. Positional Arguments

In positional arguments, values are assigned based on their position.

The order of arguments must match the order of parameters.

````python
def student(name, course):
    print("Name :", name)
    print("Course :", course)

student("Chandan", "Data Science")
````

````python
def subtract(a, b):
    print(a - b)

subtract(20, 10)
````

### Key Points

- Order is important.
- Number of arguments should match the number of parameters.

## 2. Keyword Arguments

In keyword arguments, values are passed using the parameter names.

Since each value is associated with a parameter name, the order does not matter.

````python
def student(name, course):
    print("Name :", name)
    print("Course :", course)

student(course="Data Science", name="Chandan")
````

````python
def employee(name, department):
    print(name)
    print(department)

employee(department="HR", name="Rahul")
````

### Key Points

- Order does not matter.
- Improves code readability.
- Useful when a function has many parameters.

## 3. Default Arguments

A default argument is a parameter that already has a predefined value.

If no value is passed while calling the function, the default value is used.

````python
def greet(name="Guest"):
    print("Welcome", name)

greet()
greet("Chandan")
````

````python
def calculate_interest(amount, rate=5):
    interest = amount * rate / 100
    return interest

print(calculate_interest(10000))

print(calculate_interest(10000, 8))
````

### Key Points

- Default values make functions more flexible.
- You can override the default value by passing a new value.

## 4. Variable-Length Arguments (*args)

Sometimes we don't know in advance how many arguments will be passed to a function.

In such cases, Python provides `*args`.

`*args` allows a function to accept any number of positional arguments. These arguments are stored as a tuple inside the function.

````python
def add(*numbers):
    print(numbers)

add(10, 20)
add(10, 20, 30)
add(10, 20, 30, 40)
````

### Output

```
(10, 20)
(10, 20, 30)
(10, 20, 30, 40)
```

Notice that all the values are stored as a **tuple**.

````python
def add(*numbers):
    total = 0

    for num in numbers:
        total += num

    return total

print(add(10, 20))
print(add(10, 20, 30))
print(add(10, 20, 30, 40))
````

### Real-World Example

Suppose an e-commerce website allows customers to buy any number of products.

Instead of creating separate functions for 2, 3, or 4 products, we can use `*args` to calculate the total bill for any number of items.

````python
def total_bill(*prices):
    return sum(prices)

print(total_bill(250, 400))
print(total_bill(250, 400, 800))
print(total_bill(250, 400, 800, 1200))
````

### Key Points

- `*args` accepts any number of positional arguments.
- The values are stored as a tuple.
- The name `args` is a convention; you can use any valid variable name.

### Interview Question

**Q. What is `*args` in Python?**

**Answer:**

`*args` allows a function to accept any number of positional arguments. These arguments are stored as a tuple.

## 5. Keyword Variable-Length Arguments (**kwargs)

Sometimes we don't know how many keyword arguments will be passed to a function.

In such cases, Python provides `**kwargs`.

`**kwargs` allows a function to accept any number of keyword arguments. These values are stored as a dictionary.

````python
def student(**details):
    print(details)

student(name="Chandan", age=22)
````

### Output

```
{'name': 'Chandan', 'age': 22}
```

The keyword arguments are stored as a dictionary.

````python
def student(**details):

    for key, value in details.items():
        print(key, ":", value)

student(
    name="Chandan",
    age=22,
    course="Data Science",
    city="Davanagere"
)
````

### Real-World Example

Suppose a company stores employee details.

Different employees may have different information, so `**kwargs` allows us to pass only the available details.

````python
def employee_details(**employee):
    for key, value in employee.items():
        print(key, ":", value)

employee_details(
    id=101,
    name="Rahul",
    department="HR",
    salary=50000
)
````

### Key Points

- `**kwargs` accepts any number of keyword arguments.
- The values are stored as a dictionary.
- The name `kwargs` is a convention; any valid variable name can be used.

|---------|------------|

````python
def add(a, b):
    return a + b

sum_result = add(15, 25)

print(sum_result)
````

### Key Points

The `print("End")` statement is never executed because the function exits after `return`.

## Variable Scope

The scope of a variable determines where it can be accessed in a program.

Python mainly has two types of variables based on scope:

1. Local Variable
2. Global Variable

## 1. Local Variable

A local variable is declared inside a function.

It can only be accessed within that function. Once the function finishes execution, the local variable is destroyed.

````python
def student():

    name = "Chandan"

    print("Inside Function :", name)

student()
````

### Output

```
Inside Function : Chandan
```

The variable `name` is accessible only inside the `student()` function.

````python
# def student():

#     name = "Chandan"

# student()

# print(name)


# ### Output

# ```
# NameError: name 'name' is not defined
# ```

# The variable `name` is local to the function and cannot be accessed outside it.
````

### Real-World Example

Suppose a student logs into a website.

The student's OTP is generated only during that login session.

After verification, the OTP is discarded.

This is similar to a local variable—it exists only for a specific task.

````python
def generate_otp():

    otp = 456789

    print("Generated OTP :", otp)

generate_otp()
````

## 2. Global Variable

A global variable is declared outside a function.

It can be accessed both inside and outside the function.

````python
course = "Data Science"

def display():

    print("Inside Function :", course)

display()

print("Outside Function :", course)
````

### Real-World Example

Consider a company's name.

Every department in the company uses the same company name.

Since everyone needs access to it, it can be treated like a global variable.

````python
company = "ABC Technologies"

def employee():

    print(company)

employee()

print(company)
````

### Key Points

- Declared outside the function.
- Can be accessed anywhere in the program.
- Suitable for values shared across multiple functions.

## The global Keyword

Normally, we can read a global variable inside a function.

However, to modify its value inside a function, we must use the `global` keyword.

````python
count = 0

def increment():

    global count

    count += 1

increment()

print(count)
````

````python
# count = 0

# def increment():

#     count += 1

# increment()



# ### Output

# ```
# UnboundLocalError:
# local variable 'count' referenced before assignment
# ```

# Without the `global` keyword, Python treats `count` as a local variable and raises an error.
````

## Local Variable vs Global Variable

| Local Variable | Global Variable |
|----------------|-----------------|
| Declared inside a function | Declared outside a function |
| Accessible only inside the function | Accessible throughout the program |
| Exists only during function execution | Exists until the program ends |

## Variable Shadowing

Variable shadowing occurs when a local variable has the same name as a global variable.

In such cases, the local variable takes priority inside the function, while the global variable remains unchanged outside the function.

````python
name = "Chandan"

def student():
    name = "Rahul"
    print("Inside Function :", name)

student()

print("Outside Function :", name)
````

### Key Points

- A local variable can have the same name as a global variable.
- Inside the function, the local variable is given priority.
- Outside the function, the global variable remains unchanged.

### Interview Question

**Q. What is variable shadowing?**

**Answer:**

Variable shadowing occurs when a local variable has the same name as a global variable. Inside the function, the local variable hides the global variable.

## Anonymous (Lambda) Functions

A lambda function is a small anonymous function that can have any number of arguments but only one expression.

Unlike a normal function, a lambda function is created without using the `def` keyword.

Lambda functions are commonly used for short and simple operations.

## Syntax

```python
lambda arguments : expression
```

- `lambda` → Keyword used to create an anonymous function.
- `arguments` → Input values.
- `expression` → A single expression whose result is returned automatically.
```

````python
square = lambda x: x ** 2

print(square(5))
````

````python
add = lambda a, b: a + b

print(add(10, 20))
````

### Why Use Lambda Functions?

Lambda functions are useful when a function is required only once and the logic is simple.

They help write shorter and cleaner code.

````python
numbers = [10, 20, 30, 40]

double = list(map(lambda x: x * 2, numbers))

print(double)
````

````python
marks = [35, 65, 80, 42, 91]

passed = list(filter(lambda x: x >= 40, marks))

print(passed)
````

### Real-World Example

Suppose you have a list of employee salaries and want to increase each salary by 10%.

Using a lambda function with `map()` makes this task simple.

````python
salaries = [30000, 45000, 50000]

updated_salary = list(map(lambda salary: salary * 1.10, salaries))

print(updated_salary)
````

### Lambda Function vs Normal Function

|-----------------|-----------------|
| Created using `def` | Created using `lambda` |
| Can have multiple statements | Can contain only one expression |
| Suitable for complex logic | Suitable for simple operations |
| Has a function name | Usually anonymous |

### Key Points

- Lambda functions are anonymous functions.
- They are best suited for short and simple operations.
- They automatically return the result of the expression.
- They are commonly used with `map()`, `filter()`, and `sorted()`.

## Recursive Functions

A recursive function is a function that calls itself to solve a problem.

Instead of using loops, recursion repeatedly calls the same function until a stopping condition (base case) is reached.

Recursion is useful for solving problems that can be broken down into smaller subproblems.

## How Recursion Works

Every recursive function has two important parts:

1. Base Case
   - The condition that stops the recursion.

2. Recursive Call
   - The function calls itself with a smaller or simpler input.

## Syntax

```python
def function_name(parameters):

    if base_condition:
        return value

    return function_name(smaller_problem)
```

````python
def countdown(n):

    if n == 0:
        print("Done")
        return

    print(n)

    countdown(n - 1)

countdown(5)
````

### Explanation

- The function prints the current value of `n`.
- It then calls itself with `n - 1`.
- When `n` becomes `0`, the base case is reached and the recursion stops.

## Example 1: Factorial Using Recursion

The factorial of a number is:

```
5! = 5 × 4 × 3 × 2 × 1
```

````python
def factorial(n):

    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)

print(factorial(5))
````

### Step-by-Step Execution

```
factorial(5)

↓

5 × factorial(4)

↓

5 × 4 × factorial(3)

↓

5 × 4 × 3 × factorial(2)

↓

5 × 4 × 3 × 2 × factorial(1)

↓

5 × 4 × 3 × 2 × 1

↓

120
```

## Example 2: Sum of First N Natural Numbers

````python
def sum_n(n):

    if n == 1:
        return 1

    return n + sum_n(n - 1)

print(sum_n(5))
````

## Real-World Example

Imagine opening folders inside folders on your computer.

A function can keep opening subfolders until there are no more folders left.

This process naturally follows recursion because the same task is repeated on smaller folders.

## Advantages of Recursion

- Makes code shorter for recursive problems.
- Easy to understand for tree and graph problems.
- Useful for divide-and-conquer algorithms.
- Reduces complex looping logic in some cases.

## Limitations of Recursion

- Uses more memory because each function call is stored on the call stack.
- Can be slower than loops.
- Missing a base case leads to infinite recursion.

````python
# def demo():

#     print("Hello")

#     demo()

# demo()





# ### Output

# ```
# RecursionError: maximum recursion depth exceeded
# ```

# This happens because there is no base case to stop the recursive calls.
````

## Loop vs Recursion

| Loop | Recursion |
|------|-----------|
| Uses `for` or `while` | Function calls itself |
| Usually faster | Usually slower |
| Uses less memory | Uses more memory |
| Easier for simple repetitive tasks | Better for problems with recursive structure |

## Data Science Connection

Although recursion is not used frequently in day-to-day data analysis, it is important in:

- Tree traversal
- Decision Trees
- File and folder traversal
- Graph algorithms
- Backtracking problems

## Key Points

- A recursive function calls itself.
- Every recursive function must have a base case.
- The recursive call should move towards the base case.
- Without a base case, a `RecursionError` occurs.

## Nested Functions

A nested function is a function defined inside another function.

The function defined inside is called the **inner function**, and the function containing it is called the **outer function**.

The inner function can access the variables of the outer function.

## Why Do We Need Nested Functions?

Nested functions help us:

- Organize code logically.
- Hide helper functions from the outside.
- Access variables of the outer function.
- Improve code readability.

## Syntax

```python
def outer_function():

    def inner_function():
        pass

    inner_function()

outer_function()
```

````python
def outer():

    print("Inside Outer Function")

    def inner():
        print("Inside Inner Function")

    inner()

outer()
````

### Explanation

- `outer()` is called.
- Inside `outer()`, the `inner()` function is created.
- The `inner()` function is then called.
- After the outer function finishes execution, the inner function is no longer accessible.

````python
# def outer():

#     def inner():
#         print("Hello")

# inner()

# ### Output

# ```
# NameError: name 'inner' is not defined
# ```

# The inner function exists only inside the outer function.
````

## Accessing Outer Function Variables

One advantage of nested functions is that the inner function can directly access variables of the outer function.

````python
def student():

    name = "Chandan"

    def display():
        print(name)

    display()

student()

# The inner function successfully accesses the variable `name` from the outer function
````

## Real-World Example

Suppose an online banking application first verifies the user's login.

Only after successful verification should the account balance be displayed.

The balance-checking function can be placed inside the login function because it should not be accessed directly.

````python
def login():

    print("Login Successful")

    def show_balance():
        print("Available Balance : ₹50,000")

    show_balance()

login()
````

## Returning an Inner Function

A nested function can also be returned from the outer function.

This concept is used in **closures** and **decorators**.

````python
def outer():

    def inner():
        print("Hello from Inner Function")

    return inner

result = outer()

result()
````

### Explanation

- `outer()` returns the inner function.
- The returned function is stored in `result`.
- Calling `result()` executes the inner function.

## Advantages of Nested Functions

- Improves code organization.
- Hides helper functions.
- Allows access to outer function variables.
- Forms the foundation of closures and decorators.

## Limitations

- Inner functions cannot be called directly from outside.
- Excessive nesting can reduce readability.

## Data Science Connection

However, they are important because:

## Docstrings

A docstring is a string literal used to describe the purpose and functionality of a function, class, or module.

It is written as the first statement inside the function using triple quotes (`""" """` or `''' '''`).

Docstrings improve code readability and help other developers understand the code.

## Syntax

```python
def function_name():
    """
    Description of the function.
    """
```

````python
def greet():
    """
    This function prints a welcome message.
    """
    print("Welcome to Data Science")

greet()
````

````python
print(greet.__doc__)
````

````python
def add(a, b):
    """
    Returns the sum of two numbers.
    """
    return a + b

print(add(10, 20))
print(add.__doc__)
````

## Advantages of Docstrings

- Improves readability.
- Acts as built-in documentation.
- Helps IDEs display function information.
- Useful while working in teams.

## Real-World Example

In large projects, developers may not know what every function does.

A docstring explains the function without reading the entire code.

## Data Science Connection

- NumPy

- Scikit-Learn
- TensorFlow

Almost every professional Python library documents its functions using docstrings.

# Built-in vs User-defined Functions

## Built-in Functions

Built-in functions are functions already provided by Python.

They can be used directly without defining them.

````python
numbers = [10, 20, 30, 40]

print(len(numbers))
print(max(numbers))
print(min(numbers))
print(sum(numbers))
````

## Common Built-in Functions

- print()
- input()
- len()
- max()
- min()
- sum()
- sorted()
- type()
- abs()
- round()

## User-defined Functions

User-defined functions are created by programmers using the `def` keyword.

These functions are written to solve specific problems.

````python
def square(num):
    return num ** 2

print(square(6))
````

````python
def square(num):
    return num ** 2

print(square(6))
````

## Comparison

| Built-in Functions | User-defined Functions |
|-------------------|------------------------|
| Already available | Created by programmer |
| No definition required | Defined using `def` |
| General-purpose | Problem-specific |
| Provided by Python | Written by the user |

## Real-World Example

While developing a machine learning project:

- `len()` is used to count rows.
- `max()` is used to find maximum values.
- A custom function like `calculate_accuracy()` is created by the developer.

The first two are built-in functions, while the last one is a user-defined function.

## Key Points

- Python provides many built-in functions.
- User-defined functions solve custom problems.
- Both are commonly used together in real-world applications.
