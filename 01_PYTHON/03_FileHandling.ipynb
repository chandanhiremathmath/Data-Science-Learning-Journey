# File Handling in Python

## Introduction

File handling is the process of creating, opening, reading, writing, updating, and deleting files using a programming language.

Python provides built-in functions that make file handling simple and efficient.

Instead of storing data only in memory, file handling allows data to be stored permanently on the computer.

## Why Do We Need File Handling?

Variables store data temporarily in RAM.

Once the program stops executing, all variable data is lost.

File handling helps us save data permanently so that it can be accessed even after the program ends.

Examples:

- Saving student records
- Storing employee information
- Reading datasets for Machine Learning
- Saving model predictions
- Maintaining application logs

## Real-Life Example

Consider a banking application.

When a customer deposits money, the updated balance should not disappear after closing the application.

The balance is stored in a file or database so that it is available the next time the customer logs in.

File handling helps store such information permanently.

## Types of Files

Python mainly works with two types of files.

### 1. Text Files

These files store data in human-readable form.

Examples:

- .txt
- .csv
- .json
- .py

---

### 2. Binary Files

These files store data in binary format (0s and 1s).

Examples:

- .jpg
- .png
- .mp3
- .mp4
- .pdf
- .exe

## Interview Question

### Q1. What is the difference between a text file and a binary file?

**Text File**
- Stores data as readable characters.
- Can be opened using a text editor.
- Examples: `.txt`, `.csv`, `.json`

**Binary File**
- Stores data in binary format.
- Not directly readable by humans.
- Examples: `.jpg`, `.pdf`, `.mp3`

## Why Should We Close a File?

Closing a file releases the resources used by the operating system.

If a file is not closed:

- Memory may not be released properly.
- Changes may not be saved.
- Other programs may not be able to access the file.

Therefore, always close the file after completing the operation.

# File Pointer

Whenever a file is opened, Python maintains a **file pointer**.

A file pointer is an internal cursor that keeps track of the current position in the file.

As data is read or written, the file pointer automatically moves.

Understanding the file pointer is essential because methods like `read()`, `readline()`, `seek()`, and `tell()` depend on it.

## Example

Suppose a file contains:

```
Python
Java
C++
```

Initially, the file pointer is at the beginning of the file.

```
↓
Python
Java
C++
```

After reading a few characters, the pointer moves forward automatically.

# tell()

The returned value represents the number of bytes from the beginning of the file.

# Advanced File Modes

Python provides advanced file modes by combining the basic modes with the `+` symbol.

The `+` symbol allows both **reading and writing** operations on the same file.

The advanced file modes are:

- `r+`
- `w+`
- `a+`
- `x+`

# r+ Mode

The `r+` mode opens an existing file for both reading and writing.

Features:

- File must already exist.
- Reading and writing are both allowed.
- The file pointer starts at the beginning.
- Existing data is **not deleted**.

````python
# file = open("sample.txt", "r+")

# print(file.read())

# file.seek(0)

# file.write("Hi")

# file.close()
````

### Example File Before Execution

```
Hello Python
```

### File After Execution

```
Hillo Python
```

The first two characters (`He`) are replaced by `Hi`.

Notice that `r+` **does not remove the remaining content**. It only overwrites from the current file pointer position.

## Interview Question

**Q. Does `r+` delete existing data?**

No.

It only overwrites data starting from the current file pointer position.

# w+ Mode

The `w+` mode opens a file for both reading and writing.

Features:

- Creates the file if it does not exist.
- Deletes all existing content.
- File pointer starts at the beginning.

````python
# file = open("sample.txt", "w+")

# file.write("Python File Handling")

# file.seek(0)

# print(file.read())

# file.close()
````

### Explanation

Since the file is opened in `w+` mode:

- Existing content is deleted.
- New content is written.
- `seek(0)` is required before reading because after writing, the pointer is at the end of the file.

## Interview Question

**Q. Why is `seek(0)` used after writing in `w+` mode?**

After writing, the file pointer moves to the end of the file.

Using `seek(0)` moves the pointer back to the beginning, allowing the newly written content to be read.

# a+ Mode

The `a+` mode opens a file for both appending and reading.

Features:

- Creates the file if it does not exist.
- Existing data is preserved.
- New data is always added at the end of the file.
- The file pointer is initially at the end of the file for writing.

````python
# file = open("sample.txt", "a+")

# file.write("\nWelcome")

# file.seek(0)

# print(file.read())

# file.close()
````

### Output

```
Hello Python
Welcome
```

The new text is appended to the end of the file.

`seek(0)` is used to move the pointer back to the beginning before reading.

## Interview Question

**Q. Does `a+` overwrite existing data?**

No.

It always appends new data at the end of the file.

# x+ Mode

The `x+` mode creates a new file and allows both reading and writing.

Features:

- Creates a new file.
- Raises `FileExistsError` if the file already exists.
- Supports both reading and writing.

````python
# file = open("newfile.txt", "x+")

# file.write("Python")

# file.seek(0)

# print(file.read())

# file.close()
````

### Note

Run this example only once.

If `newfile.txt` already exists, Python raises:

```python
FileExistsError
```

|------|------|-------|--------------|-----------------------|------------------|
| `r+` | ✅ | ✅ | ❌ | ❌ | Beginning |
| `w+` | ✅ | ✅ | ✅ | ✅ | Beginning |
| `a+` | ✅ | ✅ | ✅ | ❌ | End (for writing) |
| `x+` | ✅ | ✅ | ✅ | ❌ | Beginning |

# How to Choose the Correct File Mode?

Choosing the correct file mode depends on what operation you want to perform.

The following comparison can help decide which mode should be used in different situations.

## Detailed Comparison of File Modes

| Mode | Read | Write | Create File | Deletes Existing Data | File Must Exist | Pointer Position | Typical Use Case |
|------|------|-------|-------------|-----------------------|----------------|------------------|------------------|
| `r` | ✅ | ❌ | ❌ | ❌ | ✅ | Beginning | Read an existing file |
| `w` | ❌ | ✅ | ✅ | ✅ | ❌ | Beginning | Replace the entire file |
| `a` | ❌ | ✅ | ✅ | ❌ | ❌ | End | Append new data |
| `x` | ❌ | ✅ | ✅ (Only if absent) | ❌ | ❌ | Beginning | Safely create a new file |
| `r+` | ✅ | ✅ | ❌ | ❌ | ✅ | Beginning | Read and update existing content |
| `w+` | ✅ | ✅ | ✅ | ✅ | ❌ | Beginning | Start a completely new file and then read it |
| `a+` | ✅ | ✅ | ✅ | ❌ | ❌ | End (for writing) | Read existing data and append new data |
| `x+` | ✅ | ✅ | ✅ (Only if absent) | ❌ | ❌ | Beginning | Create a brand-new file with read/write access |

# Key Points

- `+` enables both reading and writing.
- `r+` preserves existing data.
- `w+` deletes existing data.
- `a+` always appends new data.
- `x+` creates a new file and raises an error if it already exists.

# with Statement (Context Manager)

## What is the `with` Statement?

The `with` statement is the recommended way to work with files in Python.

It automatically manages the file resource by opening the file and closing it after the block of code finishes, even if an exception occurs.

This reduces the chances of resource leaks and makes the code cleaner and safer.

## Syntax

```python
with open("sample.txt", "r") as file:
    # Perform file operations
```

- `open()` opens the file.
- `as file` assigns the opened file object to the variable `file`.
- The file is automatically closed when the `with` block ends.

````python
# with open("sample.txt", "r") as file:
#     content = file.read()
#     print(content)
````

### Explanation

- The file is opened in read mode.
- The contents are read and printed.
- There is no need to call `file.close()` explicitly because the `with` statement closes the file automatically.

## Why Prefer `with` Over `open()` and `close()`?

Consider the following code:

```python
file = open("sample.txt", "r")

content = file.read()

file.close()
```

If an exception occurs before `file.close()` is executed, the file may remain open.

Using the `with` statement avoids this problem because Python automatically closes the file when the block exits.

## Comparison

| Using `open()` and `close()` | Using `with` |
|-------------------------------|--------------|
| File must be closed manually | File closes automatically |
| Risk of forgetting `close()` | No such risk |
| More code | Cleaner and shorter code |
| Can leave files open if an exception occurs | Automatically handles cleanup |

## Real-World Example

Suppose you are reading a CSV file containing customer data.
