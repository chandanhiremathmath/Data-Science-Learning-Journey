# File Handling in Python

## Introduction

File handling is the process of creating, opening, reading, writing, updating, and deleting files using a programming language.

Python provides built-in functions that make file handling simple and efficient.

Instead of storing data only in memory, file handling allows data to be stored permanently on the computer.

## Why Do We Need File Handling?

Variables store data temporarily in RAM.

Once the program stops executing, all variable data is lost.

File handling helps us save data permanently so that it can be accessed even after the program ends.

Examples:

- Saving student records
- Storing employee information
- Reading datasets for Machine Learning
- Saving model predictions
- Maintaining application logs

## Real-Life Example

Consider a banking application.

When a customer deposits money, the updated balance should not disappear after closing the application.

The balance is stored in a file or database so that it is available the next time the customer logs in.

File handling helps store such information permanently.

## Types of Files

Python mainly works with two types of files.

### 1. Text Files

These files store data in human-readable form.

Examples:

- .txt
- .csv
- .json
- .py

---

### 2. Binary Files

These files store data in binary format (0s and 1s).

Examples:

- .jpg
- .png
- .mp3
- .mp4
- .pdf
- .exe

## Interview Question

### Q1. What is the difference between a text file and a binary file?

**Text File**
- Stores data as readable characters.
- Can be opened using a text editor.
- Examples: `.txt`, `.csv`, `.json`

**Binary File**
- Stores data in binary format.
- Not directly readable by humans.
- Examples: `.jpg`, `.pdf`, `.mp3`

## Opening a File

Python provides the built-in `open()` function to open a file.

### Syntax

```python
file_object = open("filename", "mode")
```

- **filename** → Name or path of the file.
- **mode** → Specifies how the file should be opened.
```

### Explanation

- `open()` opens the file.
- `"sample.txt"` is the file name.
- `"r"` means read mode.
- `file.close()` closes the file after use.

## Why Should We Close a File?

Closing a file releases the resources used by the operating system.

If a file is not closed:

- Memory may not be released properly.
- Changes may not be saved.
- Other programs may not be able to access the file.

Therefore, always close the file after completing the operation.

## File Modes

Python provides different modes to perform different operations on files.

| Mode | Description |
|------|-------------|
| `r` | Read an existing file. |
| `w` | Write to a file. Creates a new file if it doesn't exist. Overwrites existing content. |
| `a` | Append data at the end of the file. |
| `x` | Create a new file. Raises an error if the file already exists. |

## Comparison of File Modes

| Mode | File Must Exist | Creates File | Deletes Existing Data |
|------|-----------------|--------------|----------------------|
| `r` | ✅ Yes | ❌ No | ❌ No |
| `w` | ❌ No | ✅ Yes | ✅ Yes |
| `a` | ❌ No | ✅ Yes | ❌ No |
| `x` | ❌ No | ✅ Yes | ❌ No (Raises error if file exists) |
