# File Handling in Python

## Introduction

File handling is the process of creating, opening, reading, writing, updating, and deleting files using a programming language.

Python provides built-in functions that make file handling simple and efficient.

Instead of storing data only in memory, file handling allows data to be stored permanently on the computer.

## Why Do We Need File Handling?

Variables store data temporarily in RAM.

Once the program stops executing, all variable data is lost.

File handling helps us save data permanently so that it can be accessed even after the program ends.

Examples:

- Saving student records
- Storing employee information
- Reading datasets for Machine Learning
- Saving model predictions
- Maintaining application logs

## Real-Life Example

Consider a banking application.

When a customer deposits money, the updated balance should not disappear after closing the application.

The balance is stored in a file or database so that it is available the next time the customer logs in.

File handling helps store such information permanently.

## Types of Files

Python mainly works with two types of files.

### 1. Text Files

These files store data in human-readable form.

Examples:

- .txt
- .csv
- .json
- .py

---

### 2. Binary Files

These files store data in binary format (0s and 1s).

Examples:

- .jpg
- .png
- .mp3
- .mp4
- .pdf
- .exe

## Interview Question

### Q1. What is the difference between a text file and a binary file?

**Text File**
- Stores data as readable characters.
- Can be opened using a text editor.
- Examples: `.txt`, `.csv`, `.json`

**Binary File**
- Stores data in binary format.
- Not directly readable by humans.
- Examples: `.jpg`, `.pdf`, `.mp3`

### Syntax

```

- **filename** → Name or path of the file.
- **mode** → Specifies how the file should be opened.
```

## Why Should We Close a File?

Closing a file releases the resources used by the operating system.

If a file is not closed:

- Memory may not be released properly.
- Changes may not be saved.
- Other programs may not be able to access the file.

Therefore, always close the file after completing the operation.

# Reading Data from a File

## read()

The `read()` method is used to read the contents of a file.

By default, it reads the entire file and returns it as a string.

## Syntax

```python
file.read(size)
```

- `size` is optional.
- If `size` is not provided, the entire file is read.
- If `size` is provided, only that many characters are read.

````python
# file = open("sample.txt", "r")

# content = file.read()

# print(content)

# file.close()
````

### Example File (sample.txt)

```
Hello Everyone
Welcome to Python
File Handling
```

### Output

```
Hello Everyone
Welcome to Python
File Handling
```

````python
# file = open("sample.txt", "r")

# print(file.read(5))

# file.close()
````

### Output

```
Hello
```

Only the first 5 characters are read.

````python
# file = open("sample.txt", "r")

# print(file.read(5))
# print(file.read(5))

# file.close()
````

### Output

```
Hello
 Ever
```

Notice that the second `read(5)` continues reading from where the first one stopped.

Python maintains a file pointer that keeps track of the current reading position.

## Important Point

After reading the entire file once, the file pointer reaches the end of the file.

If you call `read()` again, it returns an empty string.

````python
# file = open("sample.txt", "r")

# print(file.read())

# print(file.read())

# file.close()
````

### Output

```
Hello Everyone
Welcome to Python
File Handling

''
```

The second `read()` returns an empty string because the file pointer is already at the end of the file.

## Real-World Example

Suppose you have a file containing customer details.

Using `read()`, you can load the complete file into your Python program for processing.

## Data Science Connection

The `read()` method is suitable for reading small text files such as:

- Configuration files
- Small datasets
- Log files

For large datasets, libraries like Pandas are preferred because they are more efficient.

## Interview Questions

### Q1. What does `read()` return?

It returns the contents of the file as a string.

---

### Q2. What happens if `read()` is called twice?

The second call returns the remaining content. If the entire file has already been read, it returns an empty string.

---

### Q3. What does `read(10)` do?

It reads only the first 10 characters from the current file pointer position.

# writelines()

## writelines()

The `writelines()` method is used to write multiple strings to a file.

Unlike `write()`, it accepts an iterable such as a list or tuple of strings.

**Note:** `writelines()` does not automatically add newline (`\n`) characters. If you want each string on a new line, you must include `\n` yourself.

## Syntax

```python
file.writelines(iterable)
```

- `iterable` can be a list, tuple, or any iterable containing strings.
```

````python
# file = open("sample.txt", "w")

# lines = [
#     "Python\n",
#     "Java\n",
#     "C++\n"
# ]

# file.writelines(lines)

# file.close()


# ### Output (sample.txt)

# ```
# Python
# Java
# C++
# ```
````

````python
# file = open("sample.txt", "w")

# lines = [
#     "Apple",
#     "Banana",
#     "Mango"
# ]

# file.writelines(lines)

# file.close()


# ### Output (sample.txt)

# ```
# AppleBananaMango
# ```

# Since no newline characters (`\n`) were added, all the strings are written continuously.
````

## Real-World Example

Suppose you have a list of student names.

Instead of writing each name separately using multiple `write()` calls, you can use `writelines()` to write the entire list at once.

## Data Science Connection

`writelines()` is useful for:

- Saving prediction results
- Exporting processed records
- Writing logs
- Storing multiple text entries efficiently

## Difference Between `write()` and `writelines()`

| `write()` | `writelines()` |
|-----------|----------------|
| Writes a single string | Writes multiple strings |
| Accepts only one string | Accepts an iterable of strings |
| Used for small text | Used for multiple lines |
| Returns number of characters written | Returns `None` |
| Newline must be added manually | Newline must also be added manually |

## Key Points

- `write()` writes one string.
- `writelines()` writes multiple strings.
- `writelines()` does not insert newline characters automatically.
- Add `\n` manually if each string should appear on a new line.

## Interview Questions

### Q1. What is the difference between `write()` and `writelines()`?

`write()` writes a single string, whereas `writelines()` writes multiple strings from an iterable.

---

### Q2. Does `writelines()` automatically add new lines?

No. You must include `\n` in the strings if you want each one on a separate line.

---

### Q3. What type of argument does `writelines()` accept?

It accepts an iterable such as a list or tuple containing strings.

# File Pointer

Whenever a file is opened, Python maintains a **file pointer**.

A file pointer is an internal cursor that keeps track of the current position in the file.

As data is read or written, the file pointer automatically moves.

Understanding the file pointer is essential because methods like `read()`, `readline()`, `seek()`, and `tell()` depend on it.

## Example

Suppose a file contains:

```
Python
Java
C++
```

Initially, the file pointer is at the beginning of the file.

```
↓
Python
Java
C++
```

After reading a few characters, the pointer moves forward automatically.

# tell()

The `tell()` method returns the current position of the file pointer.

The returned value represents the number of bytes from the beginning of the file.

## Syntax

```python
file.tell()
```

````python
# file = open("sample.txt", "r")

# print(file.tell())

# file.close()

### Output

# ```
# 0
# ```

# The file pointer is at the beginning of the file, so the position is `0`.
````

````python
# file = open("sample.txt", "r")

# print(file.read(6))

# print(file.tell())

# file.close()


````

### Example File (sample.txt)

```
Python
Java
C++
```

### Output

```
Python
6
```

After reading 6 characters, the file pointer moves to position `6`.

````python
# file = open("sample.txt", "r")

# file.readline()

# print(file.tell())

# file.close()
````

### Explanation

After reading the first line, the pointer moves to the beginning of the second line.

`tell()` returns the current position.

## Real-World Example

Suppose a log file contains thousands of lines.

If your program stops unexpectedly, you can store the current file pointer position using `tell()` and later continue reading from the same location.

## Data Science Connection

Although Pandas is commonly used for datasets, `tell()` is useful when processing very large text files, log files, or streaming data where tracking the current reading position is important.

## Interview Questions

### Q1. What does `tell()` return?

It returns the current position of the file pointer from the beginning of the file.

---

### Q2. Does `tell()` change the file pointer?

No. It only returns the current position.

# seek()

The `seek()` method is used to move the file pointer to a specified position.

It allows you to read or write data from any location in the file.

## Syntax

```python
file.seek(offset)
```

- `offset` specifies the position (in bytes) from the beginning of the file.

````python
# file = open("sample.txt", "r")

# print(file.read())

# print(file.read())

# file.close()
````

### Output

```
Python
Java
C++


''
```

The second `read()` returns an empty string because the file pointer is already at the end of the file.

````python
# file = open("sample.txt", "r")

# print(file.read())

# file.seek(0)

# print(file.read())

# file.close()
````

### Output

```
Python
Java
C++

Python
Java
C++
```

The `seek(0)` method moves the file pointer back to the beginning of the file, allowing the file to be read again.

````python
# file = open("sample.txt", "r")

# file.seek(7)

# print(file.read())

# file.close()
````

### Output

```
Java
C++
```

The pointer is moved to the 7th byte, so reading starts from that position.

````python
# file = open("sample.txt", "r")

# print(file.tell())

# file.seek(10)

# print(file.tell())

# file.close()
````

### Output

```
0
10
```

After calling `seek(10)`, the file pointer moves to position `10`.

## Difference Between `tell()` and `seek()`

| `tell()` | `seek()` |
|-----------|----------|
| Returns the current file pointer position | Moves the file pointer to a new position |
| Does not modify the file pointer | Changes the file pointer location |
| Returns an integer | Returns the new file position (typically 0 in text mode after success) |

## Key Points

- Every opened file has a file pointer.
- `tell()` returns the current pointer position.
- `seek()` moves the pointer to a specified position.
- `seek(0)` moves the pointer to the beginning of the file.
- Understanding the file pointer helps explain the behavior of `read()` and `readline()`.
